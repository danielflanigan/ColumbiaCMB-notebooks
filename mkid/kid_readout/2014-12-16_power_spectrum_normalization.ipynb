Check Parseval's theorem: the time series variance should equal the integral of the power spectral density.

```python
snm.fractional_fluctuation_timeseries.real.var()
```

```python
np.trapz(prr, x=f)
```

```python
np.trapz(prrh, x=fh)
```

```python
np.trapz(piih, x=fh)
```
