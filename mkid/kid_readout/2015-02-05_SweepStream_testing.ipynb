Problems and mysteries:
- For all frequencies, the raw $S_{21}$ data from the netCDF file seems to move around the resonance circle in the wrong direction; the fit data moves around the circle clockwise, which is correct. Maybe $S_{21}$ should be conjugated. This might also fix the issue of the cable delay being negative.
- In the cases where the sweep and stream seem aligned, the stream still doesn't always lie on the circle; could this just be due to 1/f?

#### Make a SweepNoiseMeasurement

```python
snm = noise_measurement.SweepNoiseMeasurement(filename,
                                              sweep_group_index=sweep_group_index,
                                              timestream_group_index=timestream_group_index,
                                              resonator_index=tg_channel_index)
```

#### Make a Stream

```python
tg_start_epoch = tg.epoch[tg_channel_index]
tg_end_epoch = tg.epoch[tg_channel_index] + tg.data_len_seconds[tg_channel_index]
tg_s21 = tg.get_data_index(tg_channel_index)
tg_frequency = tg.measurement_freq[tg_channel_index]
tg_frequency
```

```python
stream = measurement.Stream(tg_frequency, tg_s21, tg_start_epoch, tg_end_epoch)
```

```python
plt.plot(abs(stream.s21))
```

```python
plt.plot(stream.s21.real, stream.s21.imag, '.', ms=0.2)
```

#### Make a ResonatorSweep:

```python
sg.timestream_group.data.shape
```

These arrays are in sorted order:

```python
streams = []
for i in range(frequencies_per_index * channel, frequencies_per_index * (channel + 1)):
    streams.append(measurement.Stream(sg.timestream_group.measurement_freq[i],
                                      sg.timestream_group.data[i,:],
                                      sg.timestream_group.epoch[channel],
                                      sg.timestream_group.epoch[channel] + sg.timestream_group.data_len_seconds[channel]))
```

```python
sweep = measurement.ResonatorSweep(streams)
```

```python
plt.plot(sweep.frequency, abs(sweep.s21), '.')
plt.plot(stream.frequency, abs(stream.s21_mean), '.')
plt.plot(snm.sweep_freqs_MHz, abs(snm.sweep_s21), '.', ms=1)
plt.plot(snm.noise_measurement_freq_MHz, abs(snm.timestream.get_data_index(snm.timestream_index).mean()), '.', ms=1)
```

```python
fig, ax = plt.subplots()
ax.plot(stream.s21.real, stream.s21.imag, '.g', ms=0.2, alpha=0.2)
for s, m in zip(sweep.streams, sweep.s21):
    ax.plot(s.s21.real, s.s21.imag, '.k', ms=0.2, alpha=0.2)
    ax.plot(m.real, m.imag, '.r')
sg_s21 = sg.s21[frequencies_per_index * channel:frequencies_per_index * (channel + 1)]
plt.plot(sg_s21.real, sg_s21.imag, '.b', ms=1)
```
