```python
from kid_readout.measurement import basic
from kid_readout.measurement.io import nc
```

```python
ncf0 = nc.NCFile('/data/detectors/2016-05-10_122534_lo_scan.nc')
ncf1 = nc.NCFile('/data/detectors/2016-05-10_123750_lo_scan.nc')
```

```python
scan0 = ncf0.SweepArray0.as_class(basic.Scan)
scan1 = ncf1.Scan0
```

```python
fig, ax = plt.subplots()
ax.plot(1e-6 * scan0.frequency, 20*np.log10(np.abs(scan0.s21_raw_mean)), '.b')
ax.plot(1e-6 * scan1.frequency, 20*np.log10(np.abs(scan1.s21_raw_mean)), '.g')
```

```python
fig, ax = plt.subplots()
ax.set_xlim(1999, 2010)
ax.plot(1e-6 * scan0.frequency, 20*np.log10(np.abs(scan0.s21_raw_mean)), '.b')
ax.plot(1e-6 * scan1.frequency, 20*np.log10(np.abs(scan1.s21_raw_mean)), '.g')
```

```python
fig, ax = plt.subplots()
for n, sa in enumerate(scan0.stream_arrays):
    ax.plot(n, np.isnan(sa.s21_raw).sum(), '.b')
for n, sa in enumerate(scan1.stream_arrays):
    ax.plot(n, np.isnan(sa.s21_raw).sum(), '.g')
```

```python
fig, ax = plt.subplots()
for n, sa in enumerate(scan0.stream_arrays):
    ax.plot(n, sa.s21_raw.size, '.b')
for n, sa in enumerate(scan1.stream_arrays):
    ax.plot(n, sa.s21_raw.size, '.g')
```
