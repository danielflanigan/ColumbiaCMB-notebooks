```python
from scipy import signal
from kid_readout.roach import demodulator
```

```python
demod11 = demodulator.Demodulator(nfft=2**11, num_taps=8, window=signal.hamming, hardware_delay_samples=0)
```

```python
shrink = 0.8
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_xscale('symlog')
flat = 0.25
edge = 0.5
ax.axvline(-flat, color='black')
ax.axvline(flat, color='black')
ax.axvline(-edge, color='gray')
ax.axvline(edge, color='gray')
ax.plot(demod11._window_frequency, demod11._window_response)
ax.plot(shrink * demod11._window_frequency, demod11._window_response)
ax.set_xlim(-2, 2)
ax.set_ylim(1e-5, 1.5);
```
