Figure out what's up with lower middle guy.

See Glenn's 2014-10-05-emcee-resonator.ipynb and resonator.mcfit for Monte Carlo analysis of fitting.

Take noise around an IQ circle: amp noise should change and device noise should be constant in freq units; the reverse in voltage units.

See nm.resonator_model.project_s21_to_delta_freq

```python
from kid_readout.utils import readoutnc
```

```python
nm = nmons[0]
```

```python
rnc = readoutnc.ReadoutNetCDF(nm.sweep_filename)
```

```python
rnc.sweep_20141022125015.timestream_group.measurement_freq.shape[0]/16 # 16 resonators
```

```python
meas_freqs = rnc.sweep_20141022125015.timestream_group.measurement_freq[:13]
s21s = rnc.sweep_20141022125015.timestream_group.data[:13,:]
```

```python
s21s.shape
```

```python
device = []
amp = []
for k in range(13):
    proj = nm.resonator_model.project_s21_to_delta_freq(meas_freqs[k], s21s[k])
    pxx, fr = mlab.psd(proj.real / (1e6 * nm.resonator_model.f_0), NFFT=512, Fs=rnc.sweep_20141022125015.timestream_group.sample_rate[0])
    plt.loglog(fr, pxx, color=plt.cm.jet(k/13))
    device.append(pxx[(fr>80) & (fr<150)].mean())
    amp.append(pxx[fr>4e3].mean())
fig = plt.figure()
plt.semilogy(meas_freqs, device, 'o-')
plt.semilogy(meas_freqs, amp, 'x-')
```
