```python
from scipy.constants import c, h, k as k_B, pi
from scipy.special import psi
import lmfit
from physics.mkid import noroozian
from kid_readout.local import data
from kid_readout.analysis import fitter
```

```python
#bb = data.JPL_2014_May_light_blackbody_temperature()
bath = data.JPL_2014_May_light_bath_temperature()
n_kids = len(bath)
```

```python
def inverse_Q(params, T):
    return (inverse_Q_qp(params, T) +
            inverse_Q_TLS(params, T) +
            params['inverse_Q_other'].value)

def inverse_Q_qp(params, T):
    f = params['f'].value
    alpha_k = params['alpha_k'].value
    gamma = params['gamma'].value
    Delta = params['Delta'].value
    N_0 = params['N_0'].value
    S_1 = noroozian.S_1(f, T, Delta)
    n_qp = noroozian.equation_2_3(T, Delta, N_0)
    return (2 * alpha_k * gamma * S_1 * n_qp /
            (4 * N_0 * Delta))

def inverse_Q_TLS(params, T):
    F_TLS_delta_0 = params['F_TLS_delta_0'].value
    f = params['f'].value
    return F_TLS_delta_0 * np.tanh(h * f / (2 * k_B * T))

def x(params, T):
    return (x_qp(params, T) + # - x_qp(params, T_0) +
            x_TLS(params, T) + # - x_TLS(params, T_0) +
            params['x_other'].value)

def x_qp(params, T):
    f = params['f'].value
    alpha_k = params['alpha_k'].value
    gamma = params['gamma'].value
    Delta = params['Delta'].value
    N_0 = params['N_0'].value
    S_2 = noroozian.S_2(f, T, Delta)
    n_qp = noroozian.equation_2_3(T, Delta, N_0)
    return (-alpha_k * gamma * S_2 * n_qp /
            (4 * N_0 * Delta))

def x_TLS(params, T):
    F_TLS_delta_0 = params['F_TLS_delta_0'].value
    f = params['f'].value    
    return F_TLS_delta_0 / pi * (psi(0.5 + h * f / (2j * pi * k_B * T)).real - np.log(h * f / (k_B * T)))

def dual_residual(params, T, x, x_error, inverse_Q, inverse_Q_error):
    pass
```

```python
T_c = 1.5
Delta = noroozian.BCS_Delta_0(T_c)
N_0 = noroozian.N_0_Al
gamma = 1

def initial_values(f_Hz, alpha_k=0.5, F_TLS_delta_0=1e-9, x_other=0, inverse_Q_other=1e-6):
    p = lmfit.Parameters()
    p.add('f', value=f_Hz, vary=False)
    p.add('Delta', value=Delta, vary=False)
    p.add('N_0', value=N_0, vary=False)
    p.add('gamma', value=gamma, vary=False)
    p.add('alpha_k', value=alpha_k, min=0, max=1)
    p.add('F_TLS_delta_0', value=F_TLS_delta_0, min=0)
    p.add('x_other', value=x_other)
    p.add('inverse_Q_other', value=inverse_Q_other, min=0)
    return p
```

```python
x_fits = []
for n in range(n_kids):
    x_fits.append(fitter.Fitter(np.array([r.T_bath for r in bath[n]]),
                               np.array([r.f_0 / bath[n][0].f_0 - 1 for r in bath[n]]),
                               model=x, guess=lambda x, y: initial_values(1e6 * bath[n][0].f_0, F_TLS_delta_0=1e-5),
                               errors=np.array([r.result.params['f_0'].stderr / r.f_0 for r in bath[n]])))
                              
```

```python
Q_fits = []
for n in range(n_kids):
    Q_fits.append(fitter.Fitter(np.array([r.T_bath for r in bath[n]]),
                               np.array([r.Q_i**-1 for r in bath[n]]),
                               model=inverse_Q, guess=lambda x, y: initial_values(1e6 * bath[n][0].f_0)))
```
