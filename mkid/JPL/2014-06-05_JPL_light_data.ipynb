All of the data shown here are extracted from least-square fits to the model
\begin{equation}
S_{21}(f) = A e^{-2 \pi j D f} \left( 1 - \frac{Q}{Q_e} \frac{1}{1 + 2 j Q (f - f_0) / f_0} \right),
\end{equation}
where $A$ is a complex scale factor, $D$ is the cable delay, $f$ is the probe tone frequency, $f_0$ is the resonant frequency, $Q$ is the loaded resonator quality factor, and $Q_e$ is the complex coupling quality factor as described in Khalil 2012. The loaded quality factor $Q$ is related to the internal quality factor $Q_i$ and the coupling quality factor $Q_e$ by
\begin{equation}
Q^{-1} = Q_i^{-1} + \operatorname{Re} \left( Q_e^{-1} \right).
\end{equation}

# Comparison to theory

Compare the expected value of $\beta$ from BCS theory to the value obtained from the black body temperature sweep data. Assume that the gap follows the BCS relation $2 \, \Delta_0 = 3.52 \, k T_c$. In the limits
\begin{align*}
1 \gg \frac{k T}{\Delta_0}
  \approx \frac{k \, 0.2 \, \mathrm{K}}{1.76 \, k \, 1.5 \, \mathrm{K}}
  = 0.08
\end{align*}
and
\begin{align*}
1 \gg \frac{\hbar \omega}{\Delta_0}
  \approx \frac{h \, 200 \, \mathrm{MHz}}{1.76 \, k \, 1.5 \, \mathrm{K}}
  = 0.004,
\end{align*}
using approximate values for this data set, we can write
\begin{align*}
S_1(\omega, T)
  &=
  \frac{2}{\pi}
  \left( \frac{2 \Delta_0}{\pi k T} \right)^{1/2}
  \sinh \left( \frac{\hbar \omega}{2 k T} \right) \,
  K_0 \left( \frac{\hbar \omega}{2 k T} \right);
\\ S_2(\omega, T)
  &=
  1 + \left( \frac{2 \Delta_0}{\pi k T} \right)^{1/2}
  \exp \left( -\frac{\hbar \omega}{2 k T} \right) \,
  I_0 \left( \frac{\hbar \omega}{2 k T} \right).
\end{align*}
The expected value for $\beta$ is
\begin{align*}
\beta
  &=
  \frac{S_2(\omega, T)}{S_1(\omega, T)}.
\end{align*}
Using hats to denote quantities extracted from the fits to the measured data,
\begin{align*}
\hat{\beta} = -2 \frac{\delta \hat{x}}{\delta \left( \hat{Q}_i^{-1} \right)}.
\end{align*}
Since the black body temperature sweep data is highly linear, $\hat{\beta}$ does not appear to vary with applied power and we extract it by fitting a line to data like the two plots above.

```python
from physics.mkid import noroozian
```

```python
T_bath = 0.2
T_c = 1.5
ff = np.linspace(bb[0][0].f_0, bb[13][0].f_0, 1e2)
# Calculate using frequency in Hz
S_1 = noroozian.equation_2_20(1e6 * ff, T_bath, noroozian.BCS_Delta_0(T_c))
S_2 = noroozian.equation_2_21(1e6 * ff, T_bath, noroozian.BCS_Delta_0(T_c))
theory_beta = S_2 / S_1
f0 = [bb[n][0].f_0 for n in range(n_kids)]
measured_beta = []
for n in range(n_kids):
    delta_inverse_Q_i = [r.Q_i**-1 - bb[n][0].Q_i**-1 for r in bb[n]]
    delta_f_0 = [r.f_0 / bb[n][0].f_0 - 1 for r in bb[n]]
    slope, intercept = np.polyfit(delta_inverse_Q_i, delta_f_0, 1)
    measured_beta.append(-2 * slope)
```

```python
fig, ax = plt.subplots()
ax.plot(ff, theory_beta, '-g', label='expected')
ax.plot(f0, measured_beta, '.b', label='measured')
ax.plot(f0, [beta / 2 for beta in measured_beta], '.k', label='measured / 2')
ax.set_xlabel('frequency / [MHz]')
ax.set_ylabel(r'$\beta$', rotation=0)
ax.set_title(r'$T_{\mathrm{bath}} = 0.2 \, \mathrm{K}, \; T_c = 1.5 \, \mathrm{K}$')
ax.set_ylim(0, 100)
ax.legend();
```

Note that the measured points are about two times the expected points, but that the exact factor seems to decrease monotonically with frequency.

Following Jonas's review, to allow for sources of loss other than quasiparticles it might be more correct to write
$$
\beta
  =
  \frac{S_2}{S_1}
  =
  -2 \frac{\delta \hat{x}}{\delta \left( Q_{qp}^{-1} \right)}
  =
  - \frac{2}{\chi_{qp}} \frac{\delta \hat{x}}{\delta \left( \hat{Q}_i^{-1} \right)}
  =
  \frac{\hat{\beta}}{\chi_{qp}},
$$
where the hats denote experimentally measured quantities. Since $\chi_{qp} \le 1$, this goes in the wrong direction to explain the discrepancy.
