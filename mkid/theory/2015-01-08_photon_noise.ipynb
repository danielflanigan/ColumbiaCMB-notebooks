Start with Jonas's narrowband formulas for thermal radiation with a photon occupation number $n_0$  detected by a noiseless single-mode detector with quantum efficiency $\eta$ and optical bandwidth $\Delta \nu << \nu$. Unless otherwise stated, all these results assume a single polarization.

The mean detector output in units of photons per time is
$$
<d> = \eta n_0 \Delta \nu,
$$
and the variance of the mean detector output after integration time $\tau$ is
$$
\sigma^2
  =
  \frac{\Delta \nu}{\tau}
  \eta n_0 (1 + \eta n_0).
$$
We can convert these quantities to mean absorbed power by multiplying the photon number by $h \nu$:
$$
<P_A>
  =
  h \nu <d>
  =
  h \nu \eta n_0 \Delta \nu;
$$
and
$$
\sigma_A^2
  =
  \frac{h^2 \nu^2 \, \Delta \nu}{\tau}
  \eta n_0 (1 + \eta n_0).
$$
Finally, the incident power is $P_I = h \nu n_0 \, \Delta \nu = h \nu <d> / \eta$, and the uncertainty in this power is
$$
\sigma_I
  =
  \frac{h \nu \sigma}{\eta}
  =
  \frac{h \nu \, \Delta \nu}{\eta \sqrt{\tau \, \Delta \nu}}
  [\eta n_0 (1 + \eta n_0)]^{1/2},
$$
which is equations 1 and 45.

Jonas says that the noise-equivalent power is $\sigma_I$ calculated for an integration time $\tau = 0.5 \, \mathrm{s}$, which gives a 1 Hz postdetection bandwidth. If we rewrite this in terms of incident power, we obtain
$$
  \sigma_I^2 (\tau = 0.5 \, \mathrm{s})
  =
  \frac{h^2 \nu^2 \, \Delta \nu}{\eta^2 (0.5 \, \mathrm{s})}
  \eta n_0 (1 + \eta n_0)
  =
  (2 s^{-1}) h \nu \frac{(h \nu n_0 \, \Delta \nu)}{\eta}
  + (2 s^{-1}) \frac{(h \nu n_0 \, \Delta \nu)^2}{\Delta \nu},
$$
and dividing both sides by 1 s$^{-1}$, or 1 Hz, gives
$$
\mathrm{NEP}_I^2
  =
  \frac{\sigma_i^2 (\tau = 0.5 \, \mathrm{s})}{1 \, \mathrm{Hz}}
  =
  \frac{2 h \nu P_I}{\eta}
  + \frac{2 P_I^2}{\Delta \nu}.
$$
Since we derived this quantity from the uncertainty on the incident power, this is the noise-equivalent (incident) power or what is sometimes called the "optical" NEP; this is opposed to the "electrical" NEP, which is related to the uncertainty on the absorbed power. We see that in the shot noise limit the NEP is improved (decreased) by increasing the optical efficiency $\eta$, but this is not true in the wave noise limit. The Dicke equation is independent of optical efficiency.

In practical situations, the detector signal is discretely sampled. If the sample rate is $f_s$ then in time $\tau$ the number of samples is $N = \tau f_s$. Thus, $\tau = N / f_s = 1 / \delta$, where $\delta$ is the frequency resolution. The Nyquist frequency is $f_s / 2$, so the output bandwidth extends from $-f_s / 2$ to (almost) $f_s / 2$. In this band the $N$ complex frequency samples are spaced by $\delta = f_s / N = \tau^{-1}$. If the input signal is real then the positive and negative frequency components are redundant, so 

If we measure $N$ samples of time-ordered data $x(t)$ with mean $\bar{x}$ and variance $\sigma_x^2$, then the variance of the mean is $\frac{\sigma_x^2}{N}$. 

If we know the responsivity to incident power $\mathrm{d} x / \mathrm{d} P_I$, we can write the uncertainty on the measured power
$$
\hat{\sigma}_I^2
  =
  \frac{\sigma_x^2}{N (\mathrm{d} x / \mathrm{d} P_I)^2}.
$$

To relate predictions to measurements, we need to understand the fluctuations that will be produced by a given amount of absorbed power. Connect the theoretical prediction and measurement using the variance of the mean of the detector signal.

Equate this with the theoretical uncertainty:
\begin{align*}
\hat{\sigma}_I^2
  &=
  \sigma_i^2 \\
\therefore \qquad  
  \frac{\sigma_x^2}{N (\mathrm{d} x / \mathrm{d} P_I)^2}
  &=
  \tau^{-1} \left( \frac{h \nu P_I}{\eta} + \frac{P_I^2}{\Delta \nu} \right) \\
\therefore \qquad
\frac{\sigma_x^2}{f_s (\mathrm{d} x / \mathrm{d} P_I)^2}
  &=
  \frac{h \nu P_I}{\eta} + \frac{P_I^2}{\Delta \nu}.
\end{align*}
To put the left-hand side in more familiar form, let's assume that the single-sided power spectral density $S_x(f)$ of the time-ordered data consists of white noise with amplitude $W^2$ (not watts) out to the Nyquist frequency $f_s / 2$. (Situations where this is not true are discussed below.) Then, Parseval's theorem gives
$$
\hat{\sigma}_x^2
  =
  \sum \Delta f \, S_x
  =
  W^2 f_s / 2.
$$
Substitute:
\begin{align*}
\frac{W^2 f_s / 2}{f_s (\mathrm{d} x / \mathrm{d} P_I)^2}
  &=
  \frac{h \nu P_I}{\eta} + \frac{P_I^2}{\Delta \nu} \\
\therefore \qquad
\frac{W^2}{(\mathrm{d} x / \mathrm{d} P_I)^2}
  &=
  \frac{2 h \nu P_I}{\eta} + \frac{2 P_I^2}{\Delta \nu}.
\end{align*}
The left-hand side is the white noise level of the single-sided PSD of the data in units of power squared per bandwidth, and the right-hand side is the standard equation for the optical NEP.

This is the practical quantity to calculate knowing $\mathrm{d} x / \mathrm{d} P_I$. If instead we know the absorbed power $P_A = \eta P_A$ and the responsivity to absorbed power $\mathrm{d} x / \mathrm{d} P_A$, then we should use
\begin{align*}
\mathrm{NEP}_A^2
  &=
  \frac{W^2}{(\mathrm{d} x / \mathrm{d} P_A)^2}
  =
  \eta^{2} \frac{W^2}{(\mathrm{d} x / \mathrm{d} P_A)^2}
  =
  \eta^{2} \left( \frac{2 h \nu P_I}{\eta} + \frac{2 P_I^2}{\Delta \nu} \right)
  =
  2 h \nu P_A + \frac{2 P_A^2}{\Delta \nu}.
\end{align*}
Not surprisingly, this equation depends only on the absorbed power and not on the optical efficiency.

What if the noise isn't white? The measurement side of the equation is
$$
\hat{\sigma}_I^2
  =
  \frac{\sigma_x^2}{N (\mathrm{d} x / \mathrm{d} P_I)^2}.
$$
For a fixed sampling time $\tau$, this quantity is independent of sampling frequency because numerator and denominator are both proportional to the sampling frequency $f_s$. Above, we assumed that the noise was white out to the Nyquist frequency. If this is not actually the case, we could decimate the time-ordered data so that it is. Reducing the sampling frequency will reduce the variance and the number of samples by the decimation factor without changing the white noise level, which validates the above.

All detectors will have a 1/f component that becomes dominant at some low frequency. If the signal of interest will appear in the time-ordered data above the 1/f knee, then I think we're justified in using the white noise level.
