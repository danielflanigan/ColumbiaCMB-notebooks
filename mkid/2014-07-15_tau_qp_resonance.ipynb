Analyze the 872.290 MHz resonance.

```python
# Uncomment the lines below if your IPython and notebook config files do not include them.
#%pylab inline
#from __future__ import division
```

```python
f_GHz, s21_mag, s21_angle = np.loadtxt('McCarrick_2014/2014-07-15_872_MHz_resonance.s2p', unpack=True, usecols=[0, 3, 4], skiprows=23)
```

```python
s21 = s21_mag * np.exp(1j * s21_angle)
```

```python
fig, ax = plt.subplots()
ax.plot(f_GHz, 20*np.log10(s21_mag), '.')
```

```python
fig, ax = plt.subplots()
ax.plot(s21.real, s21.imag, '.')
```

```python
off_resonance = (s21_mag[0] + s21_mag[-1]) / 2
```

```python
s21_normalized = s21_mag / off_resonance
```

```python
min_index = np.argmin(s21_normalized)
```

```python
f_0 = f_GHz[min_index]
s21_min = s21_normalized[min_index]
```

```python
half_left_index = np.argmin(np.abs(s21_normalized[:min_index] - (1 - (1 - s21_min) / 2)))
half_right_index = np.argmin(np.abs(s21_normalized[min_index:] - (1 - (1 - s21_min) / 2))) + min_index
```

```python
delta_f = f_GHz[half_right_index] - f_GHz[half_left_index]
Q_r = f_0 / delta_f
Q_c = Q_r / (1 - s21_min)
Q_i = Q_r / s21_min
tau_r = (np.pi * 1e9 * delta_f)**-1 # Check this -- there may be a missing factor of two.
s21_model = 1 - Q_r * Q_c**-1 * (1 + 2j * Q_r * (f_GHz - f_0) / f_0)**-1
```
