```python
data = np.load('/data/readout/archive/2016-12-08-almn-qi-x-data.npz')
```

```python
data.keys()
```

```python
i = 1e-6 * data['qi']
x = 1e-6 * data['x']
T = data['temperature']
```

```python
x_color = 'blue'
i_color = 'brown'
```

```python
fig, i_ax = plt.subplots()
x_ax = i_ax.twinx()
x_ax.set_ylim(-30, 600)
i_ax.set_ylim(-1.5, 30)
x_ax.set_xlim(0.05, 0.21)
x_ax.plot(T, 1e6 * x, '.', color=x_color)
i_ax.plot(T, 1e6 * i, '.', color=i_color)
```

```python
fig, i_ax = plt.subplots()
x_ax = i_ax.twinx()
x_ax.set_ylim(-0.1, 1.1)
i_ax.set_ylim(-0.1, 1.1)
x_ax.set_xlim(0.05, 0.21)
x_ax.plot(T, x / x.ptp(), '.', color=x_color)
i_ax.plot(T, (i - i.min()) / (i - i.min()).ptp(), '.', color=i_color)
```
