```python
from kid_readout.measurement.io import npy
from kid_readout.measurement import plot_basic
```

```python
npd = npy.NumpyDirectory('/data/detectors/2017-03-01_153029_led_off_on.npd/')
```

```python
npd.node_names()
```

```python
sss30_off = npd.SingleSweepStream1
sss30_on = npd.SingleSweepStream3
sss36_off = npd.SingleSweepStream6
sss36_on = npd.SingleSweepStream8
```

```python
fig, ax = plt.subplots()
plot_basic.resonator_amplitude(sss30_off.resonator, ax, normalize=False,
                               sweep_mean_settings={'color': 'gray'}, model_settings={'color': 'black'})
plot_basic.resonator_amplitude(sss30_on.resonator, ax, normalize=False);
```

```python
fig, ax = plt.subplots()
plot_basic.resonator_amplitude(sss36_off.resonator, ax, normalize=False,
                               sweep_mean_settings={'color': 'gray'}, model_settings={'color': 'black'})
plot_basic.resonator_amplitude(sss36_on.resonator, ax, normalize=False);
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(sss30_off.S_frequency, sss30_off.S_xx, color='black', label='off: $S_{xx}$')
ax.plot(sss30_off.S_frequency, sss30_off.S_yy, color='gray', label='off: $S_{yy}$')
ax.plot(sss30_on.S_frequency, sss30_on.S_xx, color='red', label='on: $S_{xx}$')
ax.plot(sss30_on.S_frequency, sss30_on.S_yy, color='orange', label='on: $S_{yy}$')
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(sss36_off.S_frequency, sss36_off.S_xx, color='black', label='off: $S_{xx}$')
ax.plot(sss36_off.S_frequency, sss36_off.S_yy, color='gray', label='off: $S_{yy}$')
ax.plot(sss36_on.S_frequency, sss36_on.S_xx, color='red', label='on: $S_{xx}$')
ax.plot(sss36_on.S_frequency, sss36_on.S_yy, color='orange', label='on: $S_{yy}$')
```
