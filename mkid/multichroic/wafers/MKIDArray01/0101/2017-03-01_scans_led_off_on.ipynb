```python
from kid_readout.measurement.io import npy
```

```python
npd_off = npy.NumpyDirectory('/data/detectors/2017-03-01_142642_led_off.npd/')
npd_on = npy.NumpyDirectory('/data/detectors/2017-03-01_143103_led_on.npd/')
```

```python
sweeps_off = [npd_off.read(name) for name in npd_off.node_names() if name.startswith('SweepArray')]
sweeps_on = [npd_on.read(name) for name in npd_on.node_names() if name.startswith('SweepArray')]
```

```python
off_color = 'gray'
on_color = 'red'
```

```python
def plot_off_on(sweep_off, sweep_on, ax):
    ax.plot(sweep_off.frequency_MHz, np.abs(sweep_off.s21_point_foreground), color=off_color, alpha=1)
    ax.plot(sweep_on.frequency_MHz, np.abs(sweep_on.s21_point_foreground), color=on_color, alpha=0.5)
```

```python
fig, ax = plt.subplots()
plot_off_on(sweeps_off[11], sweeps_on[11], ax)
```
