```python
npd = npy.NumpyDirectory('/data/detectors/2017-03-01_144803_led_off_on.npd/')
```

```python
ssl36 = npd.SingleSweepStreamList1
ssl42 = npd.SingleSweepStreamList4
```

```python
x_off, q_off = ssl36.x_and_q(0)
x_on, q_on = ssl36.x_and_q(1)
```

```python
fig, ax = plt.subplots()
ax.plot(ssl36.stream_list[0].sample_time, x_off, ',', color='gray')
ax.plot(ssl36.stream_list[1].sample_time, x_on, ',', color='blue')
ax.set_ylim(0, 1e-5)
```

```python
sss36_off = basic.SingleSweepStream(sweep=ssl36.single_sweep, stream=ssl36.stream_list[0])
sss36_on = basic.SingleSweepStream(sweep=ssl36.single_sweep, stream=ssl36.stream_list[1])
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(sss36_off.S_frequency, sss36_off.S_xx, color='black', label='off: $S_{xx}$')
ax.plot(sss36_off.S_frequency, sss36_off.S_yy, color='gray', label='off: $S_{yy}$')
ax.plot(sss36_on.S_frequency, sss36_on.S_xx, color='red', label='on: $S_{xx}$')
ax.plot(sss36_on.S_frequency, sss36_on.S_yy, color='orange', label='on: $S_{yy}$')
```
