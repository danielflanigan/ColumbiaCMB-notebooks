Zmuidzinas equation 60 for the quasiparticle density is
\begin{equation}
n_{qp}(T, \Gamma)
  =
  n^*
  \left[
  \left(
  \left( 1 + \frac{n_{th}(T)}{n^*} \right)^2
  + \frac{2 \Gamma \tau_{\mathrm{max}}}{V n^*}
  \right)^{1/2}
  - 1 \right],
\end{equation}
where $n_{th}(T)$ is the thermal density, $\Gamma(P_a, P_o)$ is the quasiparticle generation rate due to absorbed optical photons and readout photons.

Try the same fit as on 2014-07-15:
\begin{equation*}
\tau_{qp}
  =
  \tau_{\mathrm{max}} (1 + T_{\mathrm{load}} / T_0)^{-1/2},
\end{equation*}
where $T_0$ combines several more fundamental constants.

```python
def model_2(params, T_load):
    tau_max = params['tau_max'].value
    T_0 = params['T_0'].value
    tau_qp = tau_max * (1 + T_load / T_0)**(-1/2)
    return tau_qp

def residual_2(params, T_load, tau_data, tau_error):
    return (tau_data - model_2(params, T_load)) / tau_error

def initial_values_2(tau_max=500, T_0=3):
    p = lmfit.Parameters()
    p.add('tau_max', value=tau_max, min=0)
    p.add('T_0', value=T_0, min=0)
    return p
```

Add a variable exponent to the temperature dependence of the power:
\begin{equation*}
\tau_{qp}
  =
  \tau_{\mathrm{max}} (1 + (T_{\mathrm{load}} / T_0)^a)^{-1/2},
\end{equation*}
where $a = 1$ is the RJ limit and $a = 4$ would be the black body power integrated over all frequencies.

```python
def model_3(params, T_load):
    tau_max = params['tau_max'].value
    T_0 = params['T_0'].value
    a = params['a'].value
    tau_qp = tau_max * (1 + (T_load / T_0)**a)**(-1/2)
    return tau_qp

def residual_3(params, T_load, tau_data, tau_error):
    return (tau_data - model_3(params, T_load)) / tau_error

def initial_values_3(tau_max=500, T_0=3, a=2):
    p = lmfit.Parameters()
    p.add('tau_max', value=tau_max, min=0)
    p.add('T_0', value=T_0, min=0)
    p.add('a', value=a, min=0)
    return p
```

```python
iv2 = initial_values_2(tau_max=600)
iv2['tau_max'].vary = False
fit_2 = lmfit.minimize(residual_2, iv2, args=[T_load, tau, tau_error])
lmfit.report_fit(fit_2.params)
```

```python
iv3 = initial_values_3(tau_max=600)
iv3['tau_max'].vary = False
fit_3 = lmfit.minimize(residual_3, iv3, args=[T_load, tau, tau_error])
lmfit.report_fit(fit_3.params)
```

```python
TT = np.linspace(3, 8, 1e3)
fig, ax = plt.subplots()
ax.errorbar(T_load, tau, yerr=tau_error, fmt='.', label='data')
ax.plot(TT, model_2(fit_2.params, TT), label='fit 2')
ax.plot(TT, model_3(fit_3.params, TT), label='fit 3')
ax.set_xlabel('Load temperature / [K]')
ax.set_ylabel(r'$\tau_{qp} \, / \, [\mu \mathrm{s}]$')
ax.legend(loc='best')
ax.set_xlim(3, 8)
ax.set_ylim(0, 1000)
```

The lifetime is falling off much faster than it should if the optical generation rate is proportional to load temperature. This is consistent with the power increasing faster than expected in the RJ limit.

We subsequently decided that the slower time constant analyzed here might not be the quasiparticle lifetime because it does not affect the noise spectrum.
