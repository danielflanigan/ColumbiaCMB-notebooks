```python
time8, spike8 = np.loadtxt(os.path.join(july_11, 'T0008CH1.CSV'), unpack=True, skiprows=16, delimiter=',')
```

```python
peak = np.argmax(spike8)
```

```python
def exponential_fall(params, t):
    offset = params['offset'].value
    amplitude = params['amplitude'].value
    tau = params['tau'].value
    peak_index = params['peak_index'].value
    return offset + amplitude * np.exp(-t[peak_index:] / tau)
```

```python
def residual(params, t, data):
    peak_index = params['peak_index'].value
    return data[peak_index:] - exponential_fall(params, t)
```

```python
def guess(t, data):
    p = lmfit.Parameters()
    peak_index = np.argmax(data)
    offset = np.mean(data[-int(data.size/10):])
    p.add('peak_index', value=peak_index, vary=False)
    p.add('offset', value=offset)
    p.add('amplitude', value=data[peak_index]-offset, min=0)
    p.add('tau', value=t.ptp()/10, min=0)
    return p
```

```python
initial = guess(time8, spike8)
initial['peak_index'].value += 2e4
fit_tail = lmfit.minimize(residual, initial, args=[time8, spike8])
fit_all = lmfit.minimize(residual, guess(time8, spike8), args=[time8, spike8])
```

```python
fig, ax = plt.subplots()
ax.plot(1e3 * time8, spike8, ',')
ax.plot(1e3 * time8[fit_tail.params['peak_index'].value:], exponential_fall(fit_tail.params, time8), '-g', linewidth=0.5,
        label=r'$\tau = $' + '{:.0f}'.format(1e6 * fit_tail.params['tau'].value) + r'$\, \mu s$')
ax.plot(1e3 * time8[fit_all.params['peak_index'].value:], exponential_fall(fit_all.params, time8), '-r', linewidth=0.5,
                label=r'$\tau = $' + '{:.0f}'.format(1e6 * fit_all.params['tau'].value) + r'$\, \mu s$')
ax.legend(loc='best')
ax.set_xlabel('time / [ms]')
```
