```python
from scipy.constants import c, h, k
```

```python
f_lpf, T_lpf = np.loadtxt('/Users/daniel/Johnson+Miller/equipment/Cardiff/Cardiff_5.7_icm_LP_filter_transmittance.txt', unpack=True)
```

```python
# Upgrade this to use simulation data. Note that more of the photon flux is at the low-frequency edge of the band.
waveguide_cutoff = 127e9
```

```python
# The filter data is in GHz.
def lpf(f):
    return np.interp(f, 1e9 * f_lpf, T_lpf, left=1, right=0)
```

```python
def passband(f):
    return (f > waveguide_cutoff) * lpf(f)
```

```python
ff = np.linspace(1, 1.2e12, 1e5)
```

```python
fig, ax = plt.subplots()
ax.plot(ff / 1e9, lpf(ff))
ax.plot(ff / 1e9, passband(ff))
ax.set_xlabel('frequency / [GHz]')
ax.set_ylabel('transmission')
ax.set_ylim(-0.05, 1.05);
```

```python
fig, ax = plt.subplots()
ax.plot(ff / 1e9, lpf(ff))
ax.plot(ff / 1e9, passband(ff))
ax.set_xlabel('frequency / [GHz]')
ax.set_ylabel('transmission')
ax.set_ylim(-0.05, 1.05)
ax.set_xlim(80, 220);
```

```python
def spectral_brightness(f, T):
    return 2 * h * f**3 * c**-2 * (np.exp(h * f / (k * T)) - 1)**-1
```

```python
def number_spectral_brightness(f, T):
    return 2 * f**2 * c**-2 * (np.exp(h * f / (k * T)) - 1)**-1
```

```python
def single_mode_throughput(f):
    return c**2 * f**-2
```

```python
Ts = np.arange(3, 11)
```

```python
number_spectra = [passband(ff) * single_mode_throughput(ff) * number_spectral_brightness(ff, T) for T in Ts]
number_flux = [np.trapz(s, ff) for s in number_spectra]
```

```python
fig, ax = plt.subplots()
for s in number_spectra:
    ax.plot(ff / 1e9, s)
ax.set_xlabel('frequency / [GHz]')
ax.set_ylabel('spectral density / [s$^{-1}$ Hz$^{-1}$]')
ax.set_xlim(100, 400);
```

This is a lower bound for the flux we would expect without the filter. This should be multiplied by the number of propagating modes at each frequency.

```python
number_spectra_unfiltered = [(ff > waveguide_cutoff) * single_mode_throughput(ff) * number_spectral_brightness(ff, T) for T in Ts]
number_flux_unfiltered = [np.trapz(s, ff) for s in number_spectra_unfiltered]
```

```python
fig, ax = plt.subplots()
for s in number_spectra_unfiltered:
    ax.plot(ff / 1e9, s)
ax.set_xlabel('frequency / [GHz]')
ax.set_ylabel('spectral density / [s$^{-1}$ Hz$^{-1}$]')
ax.set_xlim(100, 400);
```

```python
fig, ax = plt.subplots()
ax.plot(Ts, number_flux, '.')
ax.plot(Ts, number_flux_unfiltered, '.')
ax.set_xlabel('temperature / [K]')
ax.set_ylabel('photon flux / [s$^{-1}$]')
#ax.set_ylim(0, 1e11)
ax.set_xlim(0, 11);
```

```python
energy_spectra = [passband(ff) * single_mode_throughput(ff) * spectral_brightness(ff, T) for T in Ts]
```

```python
fig, ax = plt.subplots()
for s in energy_spectra:
    ax.plot(ff / 1e9, 1e12 * s)
ax.set_xlabel('frequency / [GHz]')
ax.set_ylabel('spectral density / [pW Hz$^{-1}$]')
ax.set_xlim(100, 400);
```

```python
energy_flux = [np.trapz(s, ff) for s in energy_spectra]
```

```python
fig, ax = plt.subplots()
ax.plot(Ts, 1e12 * np.array(energy_flux), '.')
ax.set_xlim(0, 11)
ax.set_ylim(0, 11)
ax.set_xlabel('temperature / [K]')
ax.set_ylabel('power / [pW]')
```
