```python
import os
import lmfit
```

Fit the data to
\begin{equation*}
S(f) = A + \frac{S_0}{1 + (2 \pi f \tau)^2},
\end{equation*}
where $A$ is a white noise level that we attribute to the amplifier, $S_0$ is the white noise level at low frequency, and $\tau$ is the dominant time constant.

```python
def model(params, f):
    A = params['A'].value
    S_0 = params['S_0'].value
    tau = params['tau'].value
    S = A + S_0 * (1 + (2 * np.pi * f * tau)**2)**-1
    return S

def residual(params, f, data, errors=1):
    return (model(params, f) - data) / errors

def guess(f, data):
    p = lmfit.Parameters()
    p.add('A', value=data[-1], min=0)
    p.add('S_0', value=data[0], min=0)
    p.add('tau', value=1/f.ptp(), min=0)
    return p

def params(A, S_0, tau):
    p = lmfit.Parameters()
    p.add('A', value=A, min=0)
    p.add('S_0', value=S_0, min=0)
    p.add('tau', value=tau, min=0)
    return p
```

```python
freq = np.load('pca_freq_r0_n0.npy')
ev0 = np.load('pca_eigvals_r0_n0.npy')
ev1 = np.load('pca_eigvals_r0_n1.npy')
f_0 = 1e6 * np.load('nms0_0_f_0.npy')
Q_r = np.load('nms0_0_Q.npy')
tau_r = Q_r / (np.pi * f_0)
tau_r
```

```python
data_Hz2_per_Hz = ev1 * f_0**2
```

```python
mask = (freq > 10) & (freq < 1e4)
fit = lmfit.minimize(residual, guess(freq[mask], data_Hz2_per_Hz[mask]), args=[freq[mask], data_Hz2_per_Hz[mask]])
lmfit.report_fit(fit.params)
```
