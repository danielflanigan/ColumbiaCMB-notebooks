```python
T_package, T_load, f_r, Q, Q_i = np.loadtxt('/data/detectors/archive/2015-11-21-ybco-157.5-MHz-resonator-summary.csv', unpack=True, skiprows=1, delimiter=',')
```

```python
Q_c = (Q**-1 - Q_i**-1)**-1
```

```python
mask = T_package < 39
```

```python
plt.plot(T_package[mask], Q_c[mask], '.')
```

```python
x_color = 'blue'
i_color = 'brown'
#T_fit = np.linspace(0, 40, 1e3)
#x_fit = np.polyfit(T_package[mask], (f_r[mask] - f_r.max()) / f_r[mask], 2)
#i_fit = np.polyfit(T_package[mask], 1 / Q_i[mask], 2)

fig, x_ax = plt.subplots()
i_ax = x_ax.twinx()
x_ax.plot(T_package[mask], 1e6 * (f_r[mask] - f_r.max()) / f_r[mask],
          marker='o', markersize=2, markeredgecolor='none', linestyle='none',
          color=x_color, label='$\delta f_r / f_r$')
#x_ax.plot(T_fit, 1e6 * np.polyval(x_fit, T_fit), color=x_color)
i_ax.plot(T_package[mask], 1e6 / Q_i[mask],
          marker='s', markersize=2, markeredgecolor='none', linestyle='none',
          color=i_color, label='$Q_i^{-1}$')
#i_ax.plot(T_fit, 1e6 * np.polyval(i_fit, T_fit), color=i_color)
#i_ax.plot(T_package[mask], 1e6 / Q_c[mask], '.k')
#i_ax.plot(T_package[mask], 1e6 / Q[mask], '.r')
x_ax.set_xlabel('temperature [K]')
x_ax.set_ylabel('$10^6 \delta f_r / f_r$', color=x_color)
i_ax.set_ylabel('$10^6 Q_i^{-1}$', color=i_color)
x_ax.set_ylim(-40e3, 2e3)
i_ax.set_ylim(-4, 80)
x_ax.legend(loc='lower left', numpoints=3,)
i_ax.legend(loc='lower center', numpoints=3)
fig.savefig('YBCO_x_and_iQi_vs_T.pdf', bbox_inches='tight')
```
