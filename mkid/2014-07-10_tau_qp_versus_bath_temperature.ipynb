These are the calculated quasiparticle lifetimes, assuming a single time constant. The scope measured rise and fall times are the times between the 10% and 90% levels of the total excursion.

```python
1 / np.log(9)
```

```python
T = np.array(T_stage)
T_c = 1.46
tau_rise = 1e-6 * np.array(t_rise) / np.log(9)
tau_fall = 1e-6 * np.array(t_fall) / np.log(9)
tau_error = 1e-6 * np.array(t_error) / np.log(9)
tau_spike = 1e-6 * np.array(t_fall_spike) / np.log(9)
tau_spike_error = 1e-6 * np.array(t_fall_spike_error) / np.log(9)
```

```python
def model(params, T):
    tau_max = params['tau_max'].value
    n_star = params['n_star'].value
#    n_qp0 = params['n_qp0'].value
    n_qp = noroozian.equation_2_3(T, noroozian.BCS_Delta_0(1.46))
    tau_qp = tau_max / (1 + n_qp / n_star)
#    tau_qp = tau_max / (1 + (n_qp + n_qp0) / n_star)
    return tau_qp
```

```python
def residual(params, T, data):
    return data - model(params, T)
```

```python
def initial_values():
    p = lmfit.Parameters()
    p.add('n_star', value=100e18, min=0)
    p.add('tau_max', value=400e-6, min=0)
# The right way to do this is to incorporate a constant qp generation rate and add that to the thermal rate.
#    p.add('n_qp0', value=1e18, min=0)
    return p
```

```python
fig, ax = plt.subplots()
ax.plot(T / T_c, 1e6 * tau_rise, '.g', label='rise')
ax.plot(T / T_c, 1e6 * tau_fall, '.b', label='fall')
ax.set_xlabel('stage temperature / [K]')
ax.set_ylabel(r'$\tau_{qp} \, / \, [\mu \mathrm{s}]$')
ax.legend(loc='best')
ax.set_ylim(0, 450)
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(T_ADR, 1e6 * tau_rise, '.g', label='rise')
ax.plot(T_ADR, 1e6 * tau_fall, '.b', label='fall')
ax.set_xlabel('ADR temperature / [K]')
ax.set_ylabel(r'$\tau_{qp} \, / \, [\mu \mathrm{s}]$')
ax.legend(loc='best')
ax.set_ylim(0, 450)
ax.grid()
```

```python
fit = lmfit.minimize(residual, initial_values(), args=[T, tau_fall])
```
