```python
def load_csv(filename):
    with open(filename) as f:
        header = [f.readline()]
        while header[-1].split(',')[0] != 'Firmware Version':
            header.append(f.readline())
        columns = ['TIME', header[6].split(',')[1]]
    data = np.loadtxt(filename, skiprows=len(header), delimiter=',', usecols=(3, 4),
                    dtype=[(column, np.float) for column in columns]).view(np.recarray)
    return data, header
```
