```python
def old_normalize_f0(group):
    mask = ((group.optical_load=='dark') & (group.sweep_primary_load_temperature < 4.2) & 
                         (group.atten==39.) & (group.f_0_err/group.f_0 < 1e-6))
    group['f_0_max'] = group[mask]['f_0'].max()
    group['frac_f0'] = (group['f_0_max'] - group['f_0']) / group['f_0_max']
    return group

def old_compute_response(group):
    mask = ((group.optical_load=='light') & (group.sweep_primary_load_temperature < 12) & 
                         (group.atten==39.) & (group.f_0_err/group.f_0 < 1e-6) &
                         (group.timestream_duration>1) & (abs(group.sweep_primary_package_temperature - 0.2) < 0.02))
    y = group[mask].frac_f0
    x = group[mask].sweep_primary_load_temperature
    slope,offset = np.polyfit(x,y,1)
    group['ppm_per_K'] = slope*1e6
    group['ppm_per_K_offset'] = offset*1e6
    return group
```

```python
old = old.groupby(old.resonator_index).apply(old_normalize_f0)
old = old.groupby(old.resonator_index).apply(old_compute_response)
```

```python
f,axs = plt.subplots(nrows=5,ncols=4,figsize=(6,6))
axs = axs.flatten()
for ax,(name,group) in zip(axs,old.groupby(old.resonator_id)):
    ax.plot(group.sweep_primary_load_temperature,group.frac_f0*1e6,',')
    mask = ((group.optical_load=='light') & (group.sweep_primary_load_temperature < 12) & 
                         (group.atten==33.) & (group.f_0_err/group.f_0 < 1e-6) &
                         (group.timestream_duration>1) & (abs(group.sweep_primary_package_temperature - 0.2) < 0.02))
    ax.plot(group[mask].sweep_primary_load_temperature, group[mask].frac_f0*1e6, 'o')
    x = np.linspace(0,12,100)
    ax.plot(x,np.polyval(np.array([group.ppm_per_K.iat[0],group.ppm_per_K_offset.iat[0]]),x))
    ax.set_ylim(0,500)
    ax.set_xlim(0, 12)
    ax.text(0.9,0.1,('%.1f' % group.ppm_per_K.iat[0]),transform=ax.transAxes,va='bottom',ha='right')
```
