```python
plt.plot(broad.sweep_epoch - broad.sweep_epoch.min(), broad.sweep_primary_package_temperature, '.b', ms=1)
plt.plot(narrow.sweep_epoch - narrow.sweep_epoch.min(), narrow.sweep_primary_package_temperature, '.g', ms=1)
```

```python
plt.semilogx(broad.zbd_power, broad.sweep_primary_package_temperature, '.b', ms=1)
plt.semilogx(narrow.zbd_power, narrow.sweep_primary_package_temperature, '.g', ms=1)
```

Examine the source-off data for systematic problems. Reference all frequencies to the same point for each channel.

```python
for channel in channels:
    broad_mask = ((broad.channel==channel) & (broad.timestream_modulation_duty_cycle==1))
    narrow_mask = ((narrow.channel==channel) & (narrow.timestream_modulation_duty_cycle==1))
    plt.semilogx(1e6 * broad[broad_mask].zbd_power,
                 1e6 * (broad[broad_mask].f_0 - broad[broad_mask].f_0.max()), '.b', ms=1)
    plt.semilogx(1e6 * narrow[narrow_mask].zbd_power,
                 1e6 * (narrow[narrow_mask].f_0 - broad[broad_mask].f_0.max()), '.g', ms=1)
```
