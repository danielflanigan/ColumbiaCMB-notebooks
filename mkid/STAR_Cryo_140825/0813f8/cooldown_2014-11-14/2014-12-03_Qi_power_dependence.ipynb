```python
from kid_readout.analysis import noise_measurement, resonator, plot_resonator
plt.rcParams['font.size'] = 11
```

```python
f_100_41 = '/home/data2/2014-11-26_133127_41_dB.nc'
f_100_38 = '/home/data2/2014-11-26_133552_38_dB.nc'
f_100_35 = '/home/data2/2014-11-26_134016_35_dB.nc'
f_100_32 = '/home/data2/2014-11-26_134441_32_dB.nc'
```

```python
s = []
for n in range(16):
    s.append({41: noise_measurement.SweepNoiseMeasurement(f_100_41, resonator_index=n),
              38: noise_measurement.SweepNoiseMeasurement(f_100_38, resonator_index=n),
              35: noise_measurement.SweepNoiseMeasurement(f_100_35, resonator_index=n),
              32: noise_measurement.SweepNoiseMeasurement(f_100_32, resonator_index=n)})
```

```python
attens = sorted(s[0].keys())
```

```python
iQis = []
for n in range(16):
    iQis.append([1/s[n][atten].Q_i for atten in attens])
```

```python
for n in range(16):
    plt.plot(attens, [1e6 * iQi for iQi in iQis[n]], '-b')
plt.xlabel('DAC attenuation')
plt.ylabel(r'$10^6 Q_i^{-1}$')
plt.xlim(31, 42)
```

```python
Qcs = []
for n in range(16):
    Qcs.append([s[n][atten].fit_params['Q_e_real'].value for atten in attens])
```

```python
for n in range(16):
    plt.plot(attens, Qcs[n], '-b')
plt.xlabel('DAC attenuation')
plt.ylabel(r'$Q_c$')
plt.xlim(31, 42)
```
