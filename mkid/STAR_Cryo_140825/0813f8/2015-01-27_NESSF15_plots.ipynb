```python
dark_channels = (0, 2, 4, 10)
bad_channels = (14,)
best = (29, 32, 29, 35, 29, 29, 32, 32, 32, 32, 32, 32, 23, 32, 29, 32)
```

```python
mask = ((source.timestream_modulation_duty_cycle==0))

for channel, group in source[mask].groupby('channel'):
    group = group[group.atten==best[channel]]
    group.sort('zbd_power', inplace=True)
    if channel in dark_channels:
        color = 'black'
    else:
        color= 'red'
    plt.loglog(1e6 * group.zbd_power, group.NEP2_zbd_fit_device**(1/2),
               '-', color=color, alpha=0.3)
```

```python
min_NEP_over_P = 1.8e-5
bandwidth = 2 / min_NEP_over_P**2
1e-9 * bandwidth  # in GHz
```

```python
from scipy.constants import h, k as k_B
nu = 155e9
#bandwidth = 13e9

coupler_20_dB = 1e-2  # Nominal value; probably solid
waveguide_loss = 1e-1  # Handwaving
eccosorb_transmission = 0.265  # Brad's calculation for 0.25"; solid
dilution = 3.5e-3  # Pure handwaving: need to do GRASP; this will vary between detectors.
zbd_to_detector = coupler_20_dB * waveguide_loss * eccosorb_transmission * dilution
print(zbd_to_detector)

# The optical efficiency for the dominant polarization, from sims.
eta = 0.7

PP = 1e-6 * np.logspace(-4, 2, 1e3)  # in watts
P_incident = zbd_to_detector * PP
```

```python
T_c = 1.39
Delta = 1.76 * k_B * T_c
recombination_noise = 2 * 2 * Delta * P_incident / eta
shot_noise = 2 * h * nu * P_incident / eta
wave_noise = 2 * P_incident**2 / bandwidth
constant = np.ones_like(P_incident) * 3e-17**2
```

```python
plt.rcParams['figure.facecolor'] = 'gray'
plt.rcParams['figure.subplot.bottom'] = 0.19

mask = ((source.timestream_modulation_duty_cycle==0))

plt.loglog(1e6 * PP, recombination_noise**(1/2), '--y', label='recombination')
plt.loglog(1e6 * PP, wave_noise**(1/2), '--b', label='wave')
plt.loglog(1e6 * PP, shot_noise**(1/2), '--g', label='shot')
plt.loglog(1e6 * PP, constant**(1/2), '--k', label='constant')
plt.loglog(1e6 * PP, (recombination_noise + shot_noise + wave_noise + constant)**(1/2),
           '-r', label='sum')

for channel, group in source[mask].groupby('channel'):
    group = group[group.atten==best[channel]]
    group.sort('zbd_power', inplace=True)
    if channel not in dark_channels and channel not in bad_channels:
        plt.loglog(1e12 * zbd_to_detector * group.zbd_power,
                   zbd_to_detector * group.NEP2_zbd_fit_device**(1/2), '-k', alpha=0.3)

plt.ylim(1e-17, 1e-15)
plt.xlabel('incident power [pW]')
plt.ylabel('$\mathrm{NEP} \; [\mathrm{W} \, \mathrm{Hz}^{-1/2}$]')
plt.legend(loc='best', fontsize='x-small')

#plt.savefig("NEP_work-in-progress.pdf")
```

```python
plt.loglog(1e6 * PP, recombination_noise**(1/2) / P_incident, '--y', label='recombination')
plt.loglog(1e6 * PP, wave_noise**(1/2) / P_incident, '--b', label='wave')
plt.loglog(1e6 * PP, shot_noise**(1/2) / P_incident, '--g', label='shot')
plt.loglog(1e6 * PP, constant**(1/2) / P_incident, '--k', label='constant')
plt.loglog(1e6 * PP, (recombination_noise + shot_noise + wave_noise + constant)**(1/2) / P_incident, '-r', label='sum')

mask = ((source.timestream_modulation_duty_cycle==0))

for channel, group in source[mask].groupby('channel'):
    group = group[group.atten==best[channel]]
    group.sort('zbd_power', inplace=True)
    if channel not in dark_channels and channel not in bad_channels:
        plt.loglog(1e12 * zbd_to_detector * group.zbd_power,
                   group.NEP2_zbd_fit_device**(1/2) / group.zbd_power, '-k', alpha=0.3)

plt.ylim(1e-5, 1e-4)
plt.xlim(1e-1, 1e2)
plt.xlabel('incident power $P$ [pW]')
plt.ylabel('$\mathrm{NEP} / P \; [\mathrm{Hz}^{-1/2}$]')
plt.legend(loc='best', fontsize='x-small')
```
