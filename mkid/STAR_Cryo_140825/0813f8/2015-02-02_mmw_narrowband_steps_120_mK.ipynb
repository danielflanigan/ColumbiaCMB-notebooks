```python
import os
from glob import glob
from kid_readout.analysis import archive, dataframe, plot_dataframe
from kid_readout.analysis.resources import skip5x4
plt.rcParams['font.size'] = 11
```

```python
archive_filename = '2015-02-0203_all_mmwtonestep'
full_archive_filename = os.path.join('/home/data/archive', archive_filename + '.npy')
```

```python
df = archive.build_simple_archive(glob('/home/data/pkl/2015-02-02_*_mmwtonestep.pkl') +
                                  ['/home/data/pkl/2015-02-03_145711_mmwtonestep.pkl',
                                   '/home/data/pkl/2015-02-03_152536_mmwtonestep.pkl'],
                                  archive_name=archive_filename)
# Two files are excluded because the sweep offset was too small and some resonators moved into the range of the next lowest channel.
df.shape
```

```python
df = archive.load_archive(full_archive_filename)
del df['index']
df.shape
```

```python
channel_to_location = np.array([0, 1, 2, 3, 4, 5, 8, 9, 10, 12, 13, 15, 16, 17, 18, 19])
df = dataframe.analyze(df, channel_to_location=channel_to_location)
channels = sorted(set(df.channel))
dark_channels = (0, 2, 4, 10)
df.shape
```

```python
masker = lambda g: (g.timestream_modulation_duty_cycle==0) & g.resonator_fit_good
df = dataframe.add_XI_responsivity(df, masker=masker)
df.shape
```

```python
df = dataframe.fit_XI_responsivity(df, dataframe.guess(None, None, None), masker=masker)
df.shape
```

```python
dataframe.add_NEP2(df)
df.shape
```

```python
# These get rid of some outliers, but some remain.
df.resonator_fit_good = df.resonator_fit_good & (1e6 / df.Q_i < 1e2) & (1e6 * df.zbd_power_X < 1e3)
```

```python
archive.save_archive(df, full_archive_filename)
```
