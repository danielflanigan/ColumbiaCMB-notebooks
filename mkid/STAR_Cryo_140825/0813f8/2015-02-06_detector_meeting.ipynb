```python
noise = archive.load_archive('/home/data/archive/2014_12_18_19_all_mmwnoisestep_analyzed.npy')
```

```python
channel_to_location = np.array([0, 1, 2, 3, 4, 5, 8, 9, 10, 12, 13, 15, 16, 17, 18, 19])
device_band = (20, 200)
amplifier_band = (5e3, 8e3)
```

```python
cw = dataframe.add_channel(cw)
cw = dataframe.add_location(cw, channel_to_location)
cw = archive.add_noise_summary(cw, device_band=device_band, amplifier_band=amplifier_band)
cw = archive.add_noise_fits(cw)
cw = archive.add_total_mmw_attenuator_turns(cw)
cw = dataframe.add_zbd_power(cw)
```

```python
channels = sorted(set(cw.channel))
dark_channels = (0, 2, 4, 10)
```

```python
maximum_fractional_Q_error = 0.1
maximum_fractional_f_0_error = 1e-6
minimum_iQi = 0
maximum_iQi = 1e-3  # The good values are less than 100
cw['resonator_mask'] = ((cw.f_0_err / cw.f_0 < maximum_fractional_f_0_error) &
                        (cw.Q_err / cw.Q < maximum_fractional_Q_error) &
                        (minimum_iQi < 1 / cw.Q_i) &
                        (1 / cw.Q_i < maximum_iQi))
```

```python
maximum_fractional_f_c_error = 0.1
cw['noise_mask'] = (cw.noise_fit_fc_err / cw.noise_fit_fc < maximum_fractional_f_c_error)
```

```python
cw = dataframe.add_zbd_power_responsivities(cw[cw.resonator_mask])
cw = dataframe.add_NEP2_zbd(cw)
cw = dataframe.add_NEP2_zbd_fit(cw)
```

```python
mask = (cw.atten == 26)

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(6, 6))
x_color = 'blue'
iQi_color = 'green'
fit_powers = 1e-6 * np.linspace(0, 20, 1e3)

for plot_index, (x_ax, (channel, group)) in enumerate(zip(axes.flatten(), cw[mask].groupby('channel'))):
    group = group[~group.f_0_max.isnull()].sort('zbd_power')
    x_ax.plot(1e6 * group.zbd_power,
              1e6 * group.x, '.', color=x_color)
    x_ax.plot(1e6 * fit_powers,
              1e6 * dataframe.z(dataframe.z_initial(z_0=group.P_zbd_x_fit_z_0.iloc[0],
                                                    P_0=group.P_zbd_x_fit_P_0.iloc[0],
                                                    P_star=group.P_zbd_x_fit_P_star.iloc[0]),
                                 fit_powers),
              '-', color=x_color)
    x_ax.set_ylim(-700, 10)
    x_ax.tick_params(axis='y', colors=x_color)
    # Put tick labels only on the left edge.
    if plot_index % 4 != 0:
        x_ax.set_yticklabels([])

    iQi_ax = x_ax.twinx()
    iQi_ax.plot(1e6 * group.zbd_power,
                1e6 / group.Q_i, '.', color=iQi_color)
    iQi_ax.plot(1e6 * fit_powers,
                1e6 * dataframe.z(dataframe.z_initial(z_0=group.P_zbd_iQi_fit_z_0.iloc[0],
                                                      P_0=group.P_zbd_iQi_fit_P_0.iloc[0],
                                                      P_star=group.P_zbd_iQi_fit_P_star.iloc[0]),
                                 fit_powers),
                '-', color=iQi_color)
    iQi_ax.set_ylim(0, 40)
    iQi_ax.tick_params(axis='y', colors=iQi_color)
    # Put ticks only on the right edge.
    if plot_index % 4 != 3:
        iQi_ax.set_yticklabels([])
    
    x_ax.set_xlim(-2, 20)
    x_ax.set_xticks(np.arange(0, 25, 5))
    # Put ticks only on the bottom edge.
    if plot_index < 12:
        x_ax.set_xticklabels([])
    # Add axis labels.
    if plot_index == 12:
        x_ax.set_xlabel('ZBD power [$\mu$W]')
        x_ax.set_ylabel('$10^6 x$')
    if plot_index == 15:
        iQi_ax.set_ylabel('$10^6 Q_i^{-1}$')

    x_ax.set_title('{}: {}'.format(group.channel.iloc[0], channel_to_location[group.channel.iloc[0]]))
```
