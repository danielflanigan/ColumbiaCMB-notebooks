```python
from kid_readout.utils import readoutnc
from kid_readout.utils import starcryo_temps
```

The numbers are the DAC attenuator settings:

```python
nc63 = readoutnc.ReadoutNetCDF('/home/data2/2014-11-17_220103.nc')
nc45 = readoutnc.ReadoutNetCDF('/home/data2/2014-11-18_195920.nc')
```

```python
mean63 = np.mean(abs(nc63.timestreams[0].data), axis=1)
mean45 = np.mean(abs(nc45.timestreams[0].data), axis=1)
```

```python
epoch63 = nc63.timestreams[0].epoch
epoch45 = nc45.timestreams[0].epoch
```

```python
T63, a, b, c = starcryo_temps.get_temperatures_at(epoch63)
T45, x, y, z = starcryo_temps.get_temperatures_at(epoch45)
```

```python
plt.plot(T63, mean63, label='63 dB')
plt.plot(T45, mean45, label='45 dB')
plt.legend()
```
