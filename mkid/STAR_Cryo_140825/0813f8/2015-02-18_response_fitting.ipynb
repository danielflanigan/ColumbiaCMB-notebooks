The reduced quasiparticle density is
$$
\frac{n_{qp}}{n_*}
  =
  \left( 1 + \frac{2 \tau_{\mathrm{max}}}{n_* V} \Gamma \right)^{1/2} -1,
$$
where $\Gamma = \Gamma_t + \Gamma_a + \Gamma_o$ and these are respectively the generation rates due to thermal phonons, readout photons, and absorbed optical photons. Assume that $\Gamma_t$ and $\Gamma_o$ are independent of $\Gamma_o \propto P$, where $P$ is the absorbed optical power. Then, we can define
$$
\frac{n_{qp}}{n_*}(P)
  =
  \left( 1 + \frac{P + P_0}{P_*} \right)^{1/2} - 1,
$$
If the temperature is constant, we should be able to ignore TLS effects except for the fact that $Q_i^{-1}$ should not go to zero as the absorbed power goes to 0.

Add small-signal responsivity equations here.

Assume that the small-signal equations are valid for the large-signal responsivities:
$$
\frac{f_0(P = 0)}{f_0(P)} - 1
  \equiv
  X(P)
  =
  X_0 \frac{n_{qp}}{n_*} (P)
$$
and
$$
Q_i^{-1}(P)
  \equiv
  I(P)
  =
  I_0 \frac{n_{qp}}{n_*} (P) + I_C,
$$
where $X_0$, $I_0$, and $I_C$ are constants, with $I_C$ representing non-quasiparticle loss. The inverse functions are
$$
P(X)
  =
  P_* \left( (1 + X / X_0)^2 - 1 \right) - P_0
$$
and
$$
P(I)
  =
  P_* \left( (1 + (I - I_C) / I_0)^2 - 1 \right) - P_0.
$$

```python
mask = ((broad.atten==26) &
        (broad.channel==7) &
        broad.resonator_mask &
        (broad.x < 0) &
        (broad.timestream_modulation_duty_cycle==0))

P = broad[mask].zbd_power
X_data = -np.array(broad[mask].x)
I_data = np.array(broad[mask].Q_i**-1)
X_errors = 2**1/2 * (broad[mask].f_0_err / broad[mask].f_0)  # This is an approximation
I_errors = broad[mask].Q_i**-2 * broad[mask].Q_i_err
initial = dataframe.parameters(1e-9, 1e-6, 20e-6, 2e-6, 1e-8)
initial['P_0'].vary = False
#initial['P_star'].vary = False
PP = 1e-6 * np.logspace(-4, np.log10(30), 1e3)

result = dataframe.fit(P, X_data, I_data, deepcopy(initial), X_errors=X_errors, I_errors=I_errors) #, xtol=1e-7, ftol=1e-7)
```

```python
lmfit.report_fit(result)
```

```python
lmfit.conf_interval(result)
```
