```python
def plot_sweeps(n):
    fig, ax = plt.subplots()
    r = df[df.resonator_id==n]
    for atten, atten_group in r.groupby('atten'):
        for index, row in atten_group.iterrows():
            ax.plot(row.sweep_freqs_MHz, abs(row.sweep_normalized_s21), label=str(atten))
    ax.legend(loc='best')
    ax.set_title(str(n))
    return fig, ax
```

```python
[plot_sweeps(n) for n in range(16)];
```

```python
def plot_pca_noise(n):
    fig, ax = plt.subplots()
    r = df[df.resonator_id==n]
    for atten, atten_group in r.groupby('atten'):
        for index, row in atten_group.iterrows():
            if np.all(np.isfinite(row.pca_eigvals)):
                ax.loglog(row.pca_freq, row.pca_eigvals[1], label=str(atten))
                ax.loglog(row.pca_freq, row.pca_eigvals[0])
    ax.set_xlim(1e-3, 1e4)
    ax.legend(loc='best')
    ax.set_title(str(n))
    return fig, ax
```

```python
[plot_pca_noise(n) for n in range(16)];
```
