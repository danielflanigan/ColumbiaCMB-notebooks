The narrowband formulas (equations 1 and 45) in Jonas's photon noise paper describe thermal radiation with a photon occupation number $n_0$  detected by a noiseless single-mode, single-polarization detector with optical bandwidth $\Delta \nu << \nu$. The variance of the mean of the absorbed power after 0.5 seconds of integration is
$$
\mathrm{NEP}_A^2
  =
  \mathrm{NEP}_0^2
  + 2 h \nu P_A
  + \frac{2 P_A^2}{\Delta \nu}.
$$
where $\mathrm{NEP}_0$ is the power-independent component of the detector noise. This should be compared to the measured NEP. In our case this is given by
$$
\mathrm{NEP}_A^2
  =
  \frac{S_{xx}}{(\mathrm{d} x / \mathrm{d} P_A)^2},
$$
where $S_{xx}$ is the single-sided PSD of the time-ordered data in units of detuning $x = f / f_r - 1$.

If the detector has quantum efficiency $\eta$, the corresponding uncertainty in the incident power $P_I = P_A / \eta$ is
$$
\mathrm{NEP}_I^2
  =
  \frac{\mathrm{NEP}_A^2}{\eta^2}
  =
  \frac{\mathrm{NEP}_0^2}{\eta^2}
  + \frac{2 h \nu P_I}{\eta}
  + \frac{2 P_I^2}{\Delta \nu}.
$$
The assumption here is that $\eta$ is independent of power. Conceptually, referring the NEP to the power incident on the detector is the same as referring it to the power at any other location, as long as this power is directly proportional to the absorbed power. To compare prediction to measurement, the corresponding measured quantity should be multiplied by the same constant:
$$
\mathrm{NEP}_I^2
  =
  \frac{\mathrm{NEP}_A^2}{\eta^2}
  =
  \frac{S_{xx}}{\eta^2 (\mathrm{d} x / \mathrm{d} P_A)^2}
  =
  \frac{S_{xx}}{(\mathrm{d} x / \mathrm{d} P_I)^2}.
$$
The quantities that we measure are $S_{xx}$ and the responsivity to measured ZBD power $(\mathrm{d} x / \mathrm{d} P_Z)$, where $P_Z$ is the power detected by the ZBD. As above, we assume that the power measured by the ZBD is proportional to the power absorbed by the detector, with $P_A = \eta Z P_Z$ and $Z = P_I / P_Z \sim 10^{-6}$. We can then calculate the uncertainty in the mean of the power at the ZBD
$$
\mathrm{NEP}_Z^2
  =
  \frac{S_{xx}}{(\mathrm{d} x / \mathrm{d} P_Z)^2}
$$
and compare it to the theoretical prediction:
$$
\mathrm{NEP}_Z^2
  =
  \frac{\mathrm{NEP}_A^2}{\eta^2 Z^2}
  =
  \frac{\mathrm{NEP}_0^2}{\eta^2 Z^2} +
  \frac{2 h \nu P_Z}{\eta Z}
  + \frac{2 P_Z^2}{\Delta \nu},
$$
This is a quadratic in $P_Z$, which we can fit to determine the bandwidth $\Delta \nu$ and the product of the efficiencies $\eta Z$.

Figure 6. The noise data flattens out at a higher level than the continuous-wave data.

The above broadband and narrowband NEP curves aren't directly comparable because the reference powers are different. In the narrowband case the efficiencies $\eta(\nu)$ and $Z(\nu)$ both apply to $\nu = 156 \, \mathrm{GHz}$. In the broadband case they are band-averaged efficiencies $\bar{\eta}$ and $\bar{Z}$. The design band is more than 20 GHz but sweeping the source in narrowband mode indicates that the bandwidth is more like 6 GHz, centered near 156 GHz.

To compare the broadband and narrowband curves directly, we need to reference the NEPs to the same power. Since none of the efficiencies are known we can't do this a priori. However, we can use the frequency shift as a proxy for absorbed power, which gives an equivalent ZBD broadband power for the continuous-wave data. Assuming that the broadband and narrowband powers are directly proportional, which is approximately true in practice, we can reference the narrowband NEP to this power and compare it directly to the broadband NEP. Define the proportionality constant
$$
\gamma(\nu)
  =
  \frac{\bar{\eta} \bar{Z}}{\eta(\nu) Z(\nu)}.
$$
The broadband equation is
$$
\mathrm{NEP}_{\bar{Z}}^2
  =
  \frac{S_{xx}}{(\mathrm{d} x / \mathrm{d} P_{\bar{Z}})^2}
  =
  \frac{\mathrm{NEP}_0^2}{\eta^2 Z^2} +
  \frac{2 h \nu P_Z}{\eta Z}
  + \frac{2 P_Z^2}{\Delta \nu},
$$


```python
atten = 32

fig, ax = plt.subplots()
for channel, group in cw[(cw.atten==atten) & cw.noise_mask].groupby('channel'):
    group.sort('zbd_power', inplace=True)
    ax.loglog(1e6 * group.equivalent_zbd_broadband_power, group.equivalent_zbd_broadband_NEP2**(1/2), '-', color=cw_color, alpha=0.3)

for channel, group in noise[noise.atten==atten].groupby('channel'):
    group.sort('zbd_power', inplace=True)
    ax.loglog(1e6 * group.zbd_power, group.NEP2_zbd_fit_device**(1/2), '-r', color=noise_color, alpha=0.3)

ax.set_xlim(1e-1, 1e2)
ax.set_ylim(1e-11, 1e-9)
ax.set_xlabel('ZBD power $P_Z$ [$\mu$W]')
ax.set_ylabel('$\mathrm{NEP}_Z \; [\mathrm{W} \, \mathrm{Hz}^{-1/2}]$')
ax.text(0.1, 0.9, 'CW', color=cw_color, transform=ax.transAxes)
ax.text(0.1, 0.8, 'noise', color=noise_color, transform=ax.transAxes);
```
