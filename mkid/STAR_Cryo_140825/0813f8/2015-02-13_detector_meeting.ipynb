The quantities that we measure are the fractional frequency power spectrum $S_{xx}$ and the responsivity to measured ZBD power $(dx / dP_Z)$, where $P_Z$ is the power detected by the ZBD. We can then calculate the ZBD NEP
$$
\mathrm{NEP}_Z^2
  =
  \frac{S_{xx}}{(dx / dP_Z)^2},
$$
where $S_{xx}$ is single-sided.

Let $Z$ be the ratio of incident power to power at the ZBD: $Z = P_I / P_Z \sim 10^{-6}$. Let $\eta$ be the quantum efficiency of the detector: $\eta = P_A / P_I$, with $P_A$ the absorbed power. Then,
$$
P_A = \eta Z P_Z.
$$
Starting from Jonas's narrowband formulas (equation 1 or 45) for thermal radiation with a photon occupation number $n_0$  detected by a noiseless single-mode, single-polarization detector with quantum efficiency $\eta$ and optical bandwidth $\Delta \nu << \nu$, the NEP referred to the absorbed power is
$$
NEP_A^2
  =
  \mathrm{NEP}_0^2
  + 2 h \nu P_A
  + \frac{2 P_A^2}{\Delta \nu}.
$$
where $\mathrm{NEP}_0$ is the power-independent component of the electrical NEP. We can refer this to the power detected by the ZBD, which is what we measure:
$$
\mathrm{NEP}_Z^2
  =
  \frac{NEP_A^2}{\eta^2 Z^2}
  =
  \frac{\mathrm{NEP}_0^2}{\eta^2 Z^2} +
  \frac{2 h \nu P_Z}{\eta Z}
  + \frac{2 P_Z^2}{\Delta \nu},
$$
This is a quadratic in $P_Z$, which we can fit to determine the bandwidth $\Delta \nu$ and the product of the efficiencies $\eta Z$.
