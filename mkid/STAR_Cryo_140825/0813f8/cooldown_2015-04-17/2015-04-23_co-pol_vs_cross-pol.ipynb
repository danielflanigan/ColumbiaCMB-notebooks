```python
import warnings
from kid_readout.analysis import process_mmw_response
from kid_readout.analysis.resources import skip5x4
from equipment.VDI import zbd
z = zbd.ZBD()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    nc_co = '/home/data2/2015-04-23_115903_mmw_frequency_sweep.nc'
    co = sorted([process_mmw_response.MmwResponse(nc_co, i) for i in range(16)], key=lambda mmw:mmw.resonator.f_0)
    nc_cx = '/home/data2/2015-03-03_200805_mmw.nc'
    cx = sorted([process_mmw_response.MmwResponse(nc_cx, i) for i in range(16)], key=lambda mmw:mmw.resonator.f_0)
```

```python
cx_mask = 1e-9 * cx[0].mmw_freq >= 140
```

```python
dark_channels = (0, 2, 4, 10)
co_dark_color = 'black'
co_light_color = 'red'
cx_dark_color = 'blue'
cx_light_color = 'green'
```

#### This notebook compares data taken with the polarization of the source illumination parallel to the absorber lines to data taken with the polarzation orthogonal to them. Data taken with co-polar illumination is labeled "co," and data taken with cross-polar illumination is labeled "cx."

```python
scale = 1.6

fig, ax = plt.subplots()
ax.plot(1e-9 * co[0].mmw_freq, 1e3 * co[0].zbd_voltage, color=co_light_color, label='co')
ax.plot(1e-9 * cx[0].mmw_freq[cx_mask], 1e3 * cx[0].zbd_voltage[cx_mask], color='gray', label='cx'.format(scale))
ax.plot(1e-9 * cx[0].mmw_freq[cx_mask], scale * 1e3 * cx[0].zbd_voltage[cx_mask], color=cx_light_color, label='{:.1f} cx'.format(scale))
ax.set_xlabel('source frequency [GHz]')
ax.set_ylabel('lock-in voltage [mV]')
ax.legend(loc='best', fontsize='xx-small');
```

#### Figure 1. All of the the cross-pol data is multiplied by a constant factor that is derived from this plot. This is intended to correct for differences in power incident on the detectors.

```python
fig, ax = plt.subplots()

for i, mmw in enumerate(co):
    if i in dark_channels:
        color=co_dark_color
        alpha=0.5
    else:
        color=co_light_color
        alpha=0.5
    ax.plot(1e-9 * mmw.mmw_freq, 1e6 * abs(mmw.mmw_frac_response), linestyle='-', color=color, alpha=alpha)

for i, mmw in enumerate(cx):
    if i in dark_channels:
        color=cx_dark_color
        alpha=0.5
    else:
        color=cx_light_color
        alpha=0.5
    ax.plot(1e-9 * mmw.mmw_freq, scale * 1e6 * abs(mmw.mmw_frac_response), linestyle='-', color=color, alpha=alpha)
    
ax.text(0.7, 0.9, 'co light', color=co_light_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.7, 0.8, 'co dark', color=co_dark_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.7, 0.7, '{:.1f} cx light'.format(scale), color=cx_light_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.7, 0.6, '{:.1f} cx dark'.format(scale), color=cx_dark_color, transform=ax.transAxes, fontsize='xx-small')
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.set_xlim(140, 165)
ax.set_ylim(-1, 50);
```

#### Figure 2. This plot shows all the data. The cross-pol data has been scaled by a factor that approximately equalizes the power detected by the ZBD between data sets.

```python
fig, ax = plt.subplots()

for i, mmw in enumerate(co):
    if i in dark_channels:
        color=co_dark_color
        alpha=0.5
    else:
        color=co_light_color
        alpha=0.5
    ax.semilogy(1e-9 * mmw.mmw_freq, 1e6 * abs(mmw.mmw_frac_response), linestyle='-', color=color, alpha=alpha)

for i, mmw in enumerate(cx):
    if i in dark_channels:
        color=cx_dark_color
        alpha=0.5
    else:
        color=cx_light_color
        alpha=0.5
    ax.semilogy(1e-9 * mmw.mmw_freq, scale * 1e6 * abs(mmw.mmw_frac_response), linestyle='-', color=color, alpha=alpha)
    
ax.text(0.2, 0.2, 'co light', color=co_light_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.2, 0.1, 'co dark', color=co_dark_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.5, 0.2, '{:.1f} cx light'.format(scale), color=cx_light_color, transform=ax.transAxes, fontsize='xx-small')
ax.text(0.5, 0.1, '{:.1f} cx dark'.format(scale), color=cx_dark_color, transform=ax.transAxes, fontsize='xx-small')
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.set_xlim(140, 165)
ax.set_ylim(1e-1, 5e1);
```

#### Figure 3. The same data as Figure 2 with a log vertical scale.

```python
co_light_mean = np.mean(np.vstack([abs(co[i].mmw_frac_response) for i in range(16) if i not in dark_channels]), axis=0)
co_dark_mean = np.mean(np.vstack([abs(co[i].mmw_frac_response) for i in dark_channels]), axis=0)
cx_light_mean = np.mean(np.vstack([abs(cx[i].mmw_frac_response) for i in range(16) if i not in dark_channels]), axis=0)
cx_dark_mean = np.mean(np.vstack([abs(cx[i].mmw_frac_response) for i in dark_channels]), axis=0)
```

```python
fig, ax = plt.subplots()
ax.plot(1e-9 * co[0].mmw_freq, 1e6 * co_light_mean, color=co_light_color, label='co light')
ax.plot(1e-9 * co[0].mmw_freq, 1e6 * co_dark_mean, color=co_dark_color, label='co dark')
ax.plot(1e-9 * cx[0].mmw_freq[cx_mask], scale * 1e6 * cx_light_mean[cx_mask], color=cx_light_color, label='{:.1f} cx light'.format(scale))
ax.plot(1e-9 * cx[0].mmw_freq[cx_mask], scale * 1e6 * cx_dark_mean[cx_mask], color=cx_dark_color, label='{:.1f} cx dark'.format(scale))
ax.set_ylim(-1, 40)
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.legend(loc='best', fontsize='xx-small');
```
