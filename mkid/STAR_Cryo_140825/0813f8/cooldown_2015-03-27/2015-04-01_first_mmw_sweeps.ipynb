```python
import warnings
from kid_readout.analysis import process_mmw_response
from kid_readout.analysis.resources import skip5x4
from equipment.VDI import zbd
z = zbd.ZBD()
```

```python
nc7p0 = '/home/data2/2015-04-01_121850_mmw_frequency_sweep.nc'
nc7p5 = '/home/data2/2015-04-01_170409_mmw_frequency_sweep.nc'
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    mmw7p0 = [process_mmw_response.MmwResponse(nc7p0, i) for i in range(16)]
    mmw7p5 = [process_mmw_response.MmwResponse(nc7p5, i) for i in range(16)]
```

```python
# This doesn't correct for demodulation, but it's right up to a constant of order 1.
f_mmw = mmw7p0[0].mmw_freq
P_zbd = mmw7p0[0].zbd_voltage / z.responsivity(f_mmw)
P_zbd_7p5 = mmw7p0[0].zbd_voltage / z.responsivity(mmw7p5[0].mmw_freq)
```

# Response of dark versus light pixels

```python
dark_indices = (0, 2, 4, 10)
dark_color = 'black'
light_color = 'red'
```

```python
fig, ax = plt.subplots()

for i, mmw in enumerate(mmw7p5):
    if i in dark_indices:
        color=dark_color
        alpha=1
    else:
        color=light_color
        alpha=0.3
    ax.plot(1e-9 * mmw.mmw_freq, 1e6 * mmw.mmw_frac_response, '-', color=color, alpha=alpha)
ax.set_ylim(-1, 30)
ax.text(0.1, 0.9, 'light', color=light_color, transform=ax.transAxes)
ax.text(0.1, 0.8, 'dark', color=dark_color, transform=ax.transAxes)
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.set_ylim(-2, 20);
```

Figure 1. Millimeter-wave source frequency versus fractional frequency response. This data set was collected by sweeping the source in narrowband mode. Data for all channels is collected simultaneously.

```python
dark_mean = np.mean(np.vstack([mmw7p5[i].mmw_frac_response for i in dark_indices]), axis=0)
light_mean = np.mean(np.vstack([mmw7p5[i].mmw_frac_response for i in range(16) if i not in dark_indices]), axis=0)
```

```python
fig, ax = plt.subplots()
ax.plot(1e-9 * f_mmw, 1e6 * light_mean, color=light_color, label='light mean')
ax.plot(1e-9 * f_mmw, 1e6 * dark_mean, color=dark_color, label='dark mean')
ax.plot(1e-9 * f_mmw, 1e6 * (light_mean - dark_mean), '-b', label='light - dark')
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.legend(loc='best', fontsize='x-small');
```

Figure 2. Millimeter-wave source frequency versus fractional frequency response. The lines are the mean of all dark channels, the mean of all channels not known to be dark, and the difference between these.

```python
plt.plot(1e-9 * f_mmw, light_mean / dark_mean, '-k', label = 'light / dark')
plt.ylim(0, 3);
```
