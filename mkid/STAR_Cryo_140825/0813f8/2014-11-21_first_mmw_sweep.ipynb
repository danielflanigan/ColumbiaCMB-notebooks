```python
from kid_readout.analysis import process_mmw_response, plot_resonator, plot_wafer
from kid_readout.analysis.resources import skip5x4
```

```python
nc_6p4_turns = '/home/data2/2014-11-21_120309_mmw.nc'
nc_7p0_turns = '/home/data2/2014-11-21_140609_mmw.nc'

mmw6p4 = [process_mmw_response.MmwResponse(nc_6p4_turns, n)
          for n in range(16)]
sorted6p4 = sorted(mmw6p4, key=lambda mmw: mmw.resonator.f_0)
integrated_response_6p4 = [np.trapz(1e6 * mmw.mmw_frac_response, x=1e-9*mmw.mmw_freq)
                           for mmw in sorted6p4]

mmw7p0 = [process_mmw_response.MmwResponse(nc_7p0_turns, n)
          for n in range(16)]
sorted7p0 = sorted(mmw7p0, key=lambda mmw: mmw.resonator.f_0)
integrated_response_7p0 = [np.trapz(1e6 * mmw.mmw_frac_response, x=1e-9*mmw.mmw_freq)
                           for mmw in sorted7p0]
```

```python
bandwidth_GHz = 1e-9 * mmw6p4[0].mmw_freq.ptp()
bandwidth_GHz
```

```python
plt.plot([mmw.resonator.Q_i for mmw in sorted6p4], integrated_response_6p4 / bandwidth_GHz, '.',
         label='6.4 turns')
plt.plot([mmw.resonator.Q_i for mmw in sorted7p0], integrated_response_7p0 / bandwidth_GHz, '.',
         label='7 turns')
plt.xlim(0, 4e5)
plt.ylim(0, 14)
plt.xticks(np.arange(5)*1e5)
plt.xlabel(r'$Q_i$')
plt.ylabel('average $x$ response')
plt.legend(loc='best');
```
