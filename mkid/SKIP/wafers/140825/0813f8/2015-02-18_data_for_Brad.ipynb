This is a coarse sweep across the whole band.

```python
coarse = readoutnc.ReadoutNetCDF('/home/data2/2014-11-17_160003.nc')
```

```python
f = coarse.sweeps[0].frequency
s21 = coarse.sweeps[0].s21
```

```python
glitch = abs(s21).argmin()
```

```python
d = s21.copy()
d[glitch-1:glitch+2] = s21[glitch-2]
```

```python
plt.plot(20*np.log10(abs(s21))[glitch-10:glitch+10], '.')
plt.plot(20*np.log10(abs(d))[glitch-10:glitch+10], '.')
```

```python
plt.rcParams['figure.subplot.bottom'] = 0.20
plt.plot(f, 20*np.log10(abs(d)))
plt.xlabel('frequency [MHz]')
plt.ylabel('$S_{21}$ [dB]')

plt.savefig('/home/flanigan/full_band_S21.pdf')
np.savetxt('/home/flanigan/full_band_f.txt', f)
np.savetxt('/home/flanigan/full_band_mag_s21.txt', abs(d))
```
