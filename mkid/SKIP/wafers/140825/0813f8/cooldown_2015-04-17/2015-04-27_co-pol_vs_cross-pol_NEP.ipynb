```python
atten = 32

fig, ax = plt.subplots()

for channel, group in co[co.atten==atten].groupby('channel'):
    if channel not in bad_channels and channel not in dark_channels:
        group = group.sort('zbd_power')
        ax.loglog(1e6 * group.zbd_power, group.zbd_power_NEP2_device**(1/2) / group.zbd_power, '-', color=co_color, label='co')

for channel, group in cx[cx.atten==atten].groupby('channel'):
    if channel not in bad_channels and channel not in dark_channels:
        group = group.sort('zbd_power')
        ax.loglog(1e6 * group.zbd_power, group.zbd_power_NEP2_device**(1/2) / group.zbd_power, '-', color=cx_color, label='cx')

ax.set_xlabel('ZBD power $P_Z$ [$\mu$W]')
ax.set_ylabel('$\mathrm{NEP}_Z  / P_Z \; [\mathrm{W} \, \mathrm{Hz}^{-1/2}]$')
ax.text(0.1, 0.8, 'co', color=co_color, transform=ax.transAxes)
ax.text(0.1, 0.9, 'cx', color=cx_color, transform=ax.transAxes)
ax.set_xlim(1e-2, 1e2)
ax.set_ylim(1e-5, 1e-4)

# Figure out the bandwidths
co_minimum_nep_over_p = 1.3e-5
cx_minimum_nep_over_p = 1.9e-5
ax.axhline(co_minimum_nep_over_p, color=co_color, linestyle='--')
ax.axhline(cx_minimum_nep_over_p, color=cx_color, linestyle='--');
```

Referred to the power incident on the detector,
$$
\mathrm{NEP}^2
  =
  \frac{2 h \nu P}{\eta} + \frac{2 P^2}{\Delta \nu},
$$
where $P$ is the incident power, $\eta$ is the quantum efficiency, and $\Delta \nu$ is the bandwidth. In the wave-noise limit,
$$
\frac{\mathrm{NEP}}{P}
  =
  \left(\frac{2}{\Delta \nu}\right)^{1/2},
$$
so
$$
\Delta \nu
  \ge
  2 \left[ \mathrm{min} \left( \frac{\mathrm{NEP}}{P} \right) \right]^{-2}.
$$

```python
co_bandwidth = 2 * co_minimum_nep_over_p**-2
cx_bandwidth = 2 * cx_minimum_nep_over_p**-2
print("Minimum co-pol bandwidth [GHz]: {:.0f}".format(1e-9 * co_bandwidth))
print("Approximate cross-pol bandwidth [GHz]: {:.0f}".format(1e-9 * cx_bandwidth))
```
