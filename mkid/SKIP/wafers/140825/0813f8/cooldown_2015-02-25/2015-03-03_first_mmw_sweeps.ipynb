```python
fig, ax = plt.subplots()

for i, mmw in enumerate(mmw7p0):
    if i in dark_indices:
        color=dark_color
        alpha=1
    else:
        color=light_color
        alpha=0.3
    ax.semilogy(1e-9 * mmw.mmw_freq, 1e6 * mmw.mmw_frac_response, '-', color=color, alpha=alpha)
ax.text(0.1, 0.9, 'light', color=light_color, transform=ax.transAxes)
ax.text(0.1, 0.8, 'dark', color=dark_color, transform=ax.transAxes)
ax.set_ylabel('$10^6 x$')
ax.set_xlabel('source frequency [GHz]')
ax.set_ylim(1e-1, 1e2)
ax.set_xlim(152, 165);
```

```python
dark_mean = np.mean(np.vstack([mmw7p0[i].mmw_frac_response for i in dark_indices]), axis=0)
light_mean = np.mean(np.vstack([mmw7p0[i].mmw_frac_response for i in range(16) if i not in dark_indices]), axis=0)
```

```python
fig, x_ax = plt.subplots(figsize=(3, 3))
P_ax = x_ax.twinx()

x_ax.plot(1e-9 * f_mmw, 1e6 * light_mean, color=light_color, label='light mean')
x_ax.plot(1e-9 * f_mmw, 1e6 * dark_mean, color=dark_color, label='dark mean')
x_ax.plot(1e-9 * f_mmw, 1e6 * (light_mean - dark_mean), '-b', label='light - dark')
x_ax.legend(loc='lower left', fontsize='x-small')
x_ax.set_ylabel('$10^6 x$')
x_ax.set_xlabel('source frequency [GHz]');
x_ax.set_ylim(-8, 22)
x_ax.set_xlim(140, 165)

P_ax.plot(1e-9 * f_mmw, mmw7p0[0].zbd_voltage / mmw7p0[0].zbd_voltage.mean(), 'gray', label='ZBD')
P_ax.legend(loc='lower right')
P_ax.set_ylabel('ZDB power [arbs]')
P_ax.set_ylim(-0.8, 2.2)

# This plot was sent to Peter and Phil
#fig.savefig('/home/flanigan/mmw_source_frequency_sweep.pdf', bbox_inches='tight')
```
