```python
from kid_readout.analysis import archive, dataframe
plt.rcParams['font.size'] = 11
```

```python
dbm0 = archive.build_simple_archive('/home/data/pkl/2015-03-23_192500_electrical_crosstalk_channel_0_power_0.0_dBm.pkl')
```

```python
dbm10 = archive.build_simple_archive('/home/data/pkl/2015-03-23_193741_electrical_crosstalk_channel_0_power_10.0_dBm.pkl')
```

```python
bad_channels = (12, 14)
probe_channel = 0
```

```python
dataframe.rename_f(dbm0)
dataframe.add_channel(dbm0)
dataframe.rename_f(dbm10)
dataframe.add_channel(dbm10)
```

```python
fig, ax = plt.subplots(figsize=(4, 2))
for channel, group in dbm0.groupby('channel'):
    if channel != probe_channel and channel not in bad_channels:
        ax.errorbar(group.channel - 0.15, 1e6 * (group.f_r.max() / group.f_r - 1), yerr=1e6 * group.f_r_err / group.f_r, color='b', marker='.', linestyle='None')
for channel, group in dbm10.groupby('channel'):
    if channel != probe_channel and channel not in bad_channels:
        ax.errorbar(group.channel + 0.15, 1e6 * (group.f_r.max() / group.f_r - 1), yerr=1e6 * group.f_r_err / group.f_r, color='g', marker='.', linestyle='None')
ax.set_yscale('log', nonposy='clip')
ax.set_ylim(1e-3, 1e1)
```
