```python
from kid_readout.analysis import archive
plt.rcParams['font.size'] = 11
```

```python
n120 = archive.load_archive('/home/data/archive/2015-02-1112_all_mmwnoisestep_analyzed.npy')
n200 = archive.load_archive('/home/data/archive/2015-02-1617_mmwnoisestep_analyzed.npy')
```

```python
dark_channels = (0, 2, 4, 10)
bad_channels = (12, 14)
```

The limiting NEPs seem significantly lower in the 200 mK data only at high readout power.

```python
atten = 32

for channel, group in n120[n120.atten==atten].groupby('channel'):
    if channel not in dark_channels and channel not in bad_channels:
        group.sort('zbd_power', inplace=True)
        plt.loglog(1e6 * group.zbd_power, group.NEP2_zbd_fit_device**(1/2), '-r')

for channel, group in n200[n200.atten==atten].groupby('channel'):
    if channel not in dark_channels and channel not in bad_channels:
        group.sort('zbd_power', inplace=True)
        plt.loglog(1e6 * group.zbd_power, group.NEP2_zbd_fit_device**(1/2), '-b')
```
