```python
f0_off = df0.f_0[df0.mmw_mod_duty_cycle.isnull()].values
f0_on = df0.f_0[df0.mmw_mod_duty_cycle==1].values
turns = df0.mmw_atten_total_turns[df0.mmw_mod_duty_cycle.isnull()].values
plt.semilogy(turns / 2, -1e6 * (f0_on - f0_off) / f0_off, '.')
plt.xlim(2, 10)
plt.xlabel('attenuator turns')
plt.ylabel('$10^6 x$')
```
