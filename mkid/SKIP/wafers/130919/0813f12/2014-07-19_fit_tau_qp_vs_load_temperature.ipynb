```python
import os
import pickle
from equipment.Tektronix.oscilloscope import load_csv
from notebook.library.tau_qp import single_fit, dual_fit, plot_single_fit, plot_dual_fit
```

```python
data_directory = '/Users/daniel/Johnson+Miller/KIDs/SKIP/STAR Cryo Version 0/2014-07-11_tau_qp_load'
```

```python
with open(os.path.join(data_directory, 'pulse.pkl')) as f:
    pulse = pickle.load(f)
with open(os.path.join(data_directory, 'spike.pkl')) as f:
    spike = pickle.load(f)
```

```python
pulse_delay = 10e-6
pulse_data = []
pulse_header = []
pulse_single = []
pulse_dual = []
for filename in pulse['file']:
    data, header = load_csv(os.path.join(data_directory, filename))
    pulse_data.append(data)
    pulse_header.append(header)
    pulse_single.append(single_fit(data.TIME, data.CH1, pulse['duration'] + pulse_delay))
    pulse_dual.append(dual_fit(data.TIME, data.CH1, pulse['duration'] + pulse_delay))
```

```python
spike_delay = 10e-6 # check
spike_data = []
spike_header = []
spike_single = []
spike_dual = []
for filename in spike['file']:
    data, header = load_csv(os.path.join(data_directory, filename))
    spike_data.append(data)
    spike_header.append(header)
    spike_single.append(single_fit(data.TIME, data.CH1, spike['duration'] + spike_delay))
    spike_dual.append(dual_fit(data.TIME, data.CH1, spike['duration'] + spike_delay))
```
