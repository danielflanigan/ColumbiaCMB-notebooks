```python
import os
import lmfit
```

```python
def load_csv(filename):
    with open(filename) as f:
        header = [f.readline()]
        while header[-1].split(',')[0] != 'TIME':
            header.append(f.readline())
        columns = header[-1].rstrip().split(',')
    data = np.loadtxt(filename, skiprows=len(header), delimiter=',',
                    dtype=[(column, np.float) for column in columns]).view(np.recarray)
    return data, header
```

This model is
\begin{equation*}
y(t)
  =
  A + B e^{-t / \tau}.
\end{equation*}

```python
def exponential_fall(params, t):
    offset = params['offset'].value
    amplitude = params['amplitude'].value
    tau = params['tau'].value
    start_index = params['start_index'].value
    return offset + amplitude * np.exp(-(t[start_index:] -t[start_index]) / tau)  
```

```python
def residual(params, t, data):
    start_index = params['start_index'].value
    return data[start_index:] - exponential_fall(params, t)
```

```python
def guess(t, data):
    p = lmfit.Parameters()
    peak_index = np.argmax(data)
    offset = data[-int(data.size/10):].mean()
    p.add('peak_index', value=peak_index, vary=False)
    p.add('offset_index', value=1, min=0)
    p.add('start_index', expr='peak_index + offset_index')
    p.add('offset', value=offset)
    p.add('amplitude', value=data[peak_index]-offset, min=0)
    p.add('tau', value=t.ptp()/10, min=0)
    return p
```

This is actually done by lmfit as long as the residual function is scaled by the errors:

```python
def chi_square_per_DOF(fit, t, data):
    # Estimate the standard deviation using the beginning of the trace.
    rms = data[t < 0].std()
    degrees_of_freedom = data.size - fit.params['start_index'].value
    chi_squared = np.sum(residual(fit.params, t, data)**2) / rms**2
    return chi_squared / degrees_of_freedom
```

This model is
\begin{equation*}
y(t)
  = 
  A + B_1 e^{-t / \tau_1} + B_2 e^{-t / \tau_2}.
\end{equation*}

```python
def dual_exponential_fall(params, t):
    offset = params['offset'].value
    fast_amplitude = params['fast_amplitude'].value
    fast_tau = params['fast_tau'].value
    slow_amplitude = params['slow_amplitude'].value
    slow_tau = params['slow_tau'].value
    start_index = params['start_index'].value
    return (offset +
            fast_amplitude * np.exp(-(t[start_index:] - t[start_index]) / fast_tau) +
            slow_amplitude * np.exp(-(t[start_index:] - t[start_index]) / slow_tau))
```

```python
def dual_residual(params, t, data):
    start_index = params['start_index'].value
    return data[start_index:] - dual_exponential_fall(params, t)
```

```python
def dual_guess(t, data):
    p = lmfit.Parameters()
    smoothed_data = np.convolve(np.ones(10) / 10, data-data.mean(), mode='same')
    peak_index = np.argmax(smoothed_data)
    offset = np.mean(data[-int(data.size/10):])
    p.add('peak_index', value=peak_index, vary=False)
    p.add('offset_index', value=1, min=0)
    p.add('start_index', expr='peak_index + offset_index')
    p.add('offset', value=offset, vary=False)
    p.add('fast_amplitude', value=(smoothed_data[peak_index]-offset)/2, min=0)
    p.add('fast_tau', value=t.ptp()/10, min=0)
    p.add('delta_tau', value=t.ptp()/2, min=0)
    p.add('slow_amplitude', value=(smoothed_data[peak_index]-offset)/2, min=0)
    p.add('slow_tau', expr='fast_tau + delta_tau')
    return p
```
