```python
from kid_readout.analysis.resonator import Resonator
from kid_readout.analysis.khalil import bifurcation_s21, bifurcation_guess
from kid_readout.analysis import plot_resonator
from kid_readout.utils import readoutnc
```

```python
# stage 180 mK
nc_on = readoutnc.ReadoutNetCDF('/home/data2/2014-06-17_193337_test_compressor_on.nc')
sweep_on = nc_on.sweeps[0]
r_on = []
for i in sorted(np.unique(sweep_on.index)):
    f, s21, errors = sweep_on.select_by_index(i)
    r_on.append(Resonator(f, s21, errors=None,
                          model=bifurcation_s21, guess=bifurcation_guess))
```

```python
# stage 180 mK
nc_off = readoutnc.ReadoutNetCDF('/home/data2/2014-06-17_193848_test_compressor_off.nc')
sweep_off = nc_off.sweeps[0]
r_off = []
for i in sorted(np.unique(sweep_off.index)):
    f, s21, errors = sweep_off.select_by_index(i)
    r_off.append(Resonator(f, s21, errors=None,
                            model=bifurcation_s21, guess=bifurcation_guess))
```

```python
[plot_resonator.IQ_circle(r, title=r'$a = {:.2f} \pm {:.2f}$'.format(r.a, r.result.params['a'].stderr), normalize_s21=True) for r in r_on]
```

```python
[plot_resonator.IQ_circle(r, title=r'$a = {:.2f} \pm {:.2f}$'.format(r.a, r.result.params['a'].stderr), normalize_s21=True) for r in r_off]
```
