```python
import lmfit
```

Raw data is copied from measurements on 2014-07-11.

```python
T_eccosorb = np.array([3.973, 4.475, 5.044, 5.434, 5.876, 6.664, 7.31])
tau_fall = np.array([455.11961331,  295.82774865,  259.41817959,  200.25262986, 188.87463953,  158.86519002,  144.01691264])
tau_fall_error = np.array([ 22.75598067,  45.51196133,  31.85837293, 13.6535884,  9.10239227,  11.51311934,   8.52054046])
```

Propagate the errors $\mathrm{d} \tau$:
\begin{equation*}
\left| \mathrm{d} (\tau^{-2}) \right|
  =
  \left| -2 \tau^{-3} \mathrm{d} \tau \right|.
\end{equation*}

```python
yerr = 2 * tau_fall**-3 * tau_fall_error
```

Fit to $\tau^{-2} = a T + b$.

```python
p = np.poly1d(np.polyfit(T_eccosorb, tau_fall**-2, 1, w=yerr**-1)) # Weight each point by \sigma^{-1}.
p
```

```python
fig, ax = plt.subplots()
ax.errorbar(T_eccosorb, tau_fall**-2, fmt='.',
            yerr=yerr)
ax.plot(T_eccosorb, p(T_eccosorb), '-')
ax.set_xlabel('load temperature [K]')
ax.set_ylabel(r'$\tau_{qp}^{-2} \, [\mu \mathrm{s}^{-2}]$')
```

This is pretty linear, but note that the vertical intercept is negative.

If the optical generation rate $\Gamma$ is both sufficiently large and proportional to load temperature $T_{\mathrm{load}}$,
\begin{equation*}
\tau_{qp}
  =
  \frac{\tau_{\mathrm{max}}}{\sqrt{1 + 2 \Gamma \tau_{\mathrm{max}} / N^*}}
  \approx A T_{\mathrm{load}}^{-1/2}.
\end{equation*}
Fit the data to this equation.

```python
def model(params, T_load):
    return params['A'].value * T_load**(-1/2)
```

```python
def residual(params, T_load, tau_data):
    return tau_data - model(params, T_load)
```

```python
def initial_values():
    p = lmfit.Parameters()
    p.add('A', value=1, min=0)
    return p
```

```python
fit = lmfit.minimize(residual, initial_values(), args=[T_eccosorb, tau_fall])
```

```python
fit.params
```

```python
TT = np.linspace(T_eccosorb.min(), T_eccosorb.max(), 1e3)
fig, ax = plt.subplots()
ax.errorbar(T_eccosorb, tau_fall, fmt='.', yerr=tau_fall_error)
ax.plot(TT, model(fit.params, TT), '-')
ax.set_xlabel('load temperature [K]')
ax.set_ylabel(r'$\tau_{qp} \, [\mu \mathrm{s}]$')
ax.set_ylim(0, 500)
ax.set_xlim(3, 8);
```

That doesn't work very well. Looks like we can't neglect the 1. Try fitting to
\begin{equation*}
\tau_{qp}
  =
  \tau_{\mathrm{max}} (1 + T_{\mathrm{load}} / T_0)^{-1/2},
\end{equation*}
where $T_0$ is some combination of other constants that I'm not going to try to interpret. Above, we fit to
\begin{equation*}
\tau^{-2} = a T + b,
\end{equation*}
so
\begin{equation*}
\tau = (a T + b)^{-1/2} = b^{-1/2} (1 + (a / b) T)^{-1/2}.
\end{equation*}
Comparing this to the above equation we see that the vertical intercept $b$ has to be positive for this to make any sense, but the value obtained from the linear fit is negative. I'm not sure how to interpret this.

```python
def model_2(params, T_load):
    tau_max = params['tau_max'].value
    T_0 = params['T_0'].value
    tau_qp = tau_max * (1 + T_load / T_0)**(-1/2)
    return tau_qp
```

```python
def residual_2(params, T_load, tau_data, tau_error):
    return (tau_data - model_2(params, T_load)) / tau_error
```

```python
def initial_values_2(tau_max=400, T_0=3):
    p = lmfit.Parameters()
    p.add('tau_max', value=tau_max, min=0)
    p.add('T_0', value=T_0, min=0)
    return p
```

```python
fit_2 = lmfit.minimize(residual_2, initial_values_2(), args=[T_eccosorb, tau_fall, tau_fall_error])
```

```python
fit_2.params
```

```python
TT = np.linspace(T_eccosorb.min(), T_eccosorb.max(), 1e3)
TTT = np.linspace(0, 8, 1e3)
fig, ax = plt.subplots()
ax.errorbar(T_eccosorb, tau_fall, fmt='.', yerr=tau_fall_error)
ax.plot(TT, model_2(fit_2.params, TT), '-')
#ax.plot(TTT, model_2(initial_values_2(500, -0.1), TTT), '-')
ax.set_xlabel('load temperature [K]')
ax.set_ylabel(r'$\tau_{qp} \, [\mu \mathrm{s}]$')
ax.set_ylim(0, 500)
ax.set_xlim(3, 8);
```
