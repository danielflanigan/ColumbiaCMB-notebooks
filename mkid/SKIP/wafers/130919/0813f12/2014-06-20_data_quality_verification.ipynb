```python
from kid_readout.analysis.resonator import Resonator
from kid_readout.analysis.khalil import bifurcation_s21, bifurcation_guess
from kid_readout.analysis.khalil import delayed_auto_guess, delayed_generic_s21
from kid_readout.analysis import plot_resonator
from kid_readout.utils import readoutnc
from kid_readout.analysis import noise_measurement
plt.rcParams['font.size'] = 11
```

```python
def plot_sweep_and_ts(nc):
    fig, ax = plt.subplots()
    colors = ['b', 'g', 'r', 'c']
    for i in range(4):
        ax.plot(nc.timestreams[0].data[i].real, nc.timestreams[0].data[i].imag, ',', color=colors[i])
        f, s21, err = nc.sweeps[0].select_by_index(i)
        ax.plot(s21.real, s21.imag, '.', color=colors[i])
```

```python
def fit_resonators(sweep):
    rs = []
    for i in sorted(np.unique(sweep.index)):
        f, s21, errors = sweep.select_by_index(i)
        rs.append(Resonator(f, s21, errors=None,
                            model=delayed_generic_s21, guess=delayed_auto_guess))
    return rs

def fit_resonators_bif(sweep):
    rs = []
    for i in sorted(np.unique(sweep.index)):
        f, s21, errors = sweep.select_by_index(i)
        rs.append(Resonator(f, s21, errors=None,
                            model=bifurcation_s21, guess=bifurcation_guess))
    return rs
```
