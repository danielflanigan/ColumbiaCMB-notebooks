```python
from kid_readout.utils import dataset
```

```python
filename = '/home/data2/2014-08-05_134641.nc'
ds = dataset.Dataset(filename)
```

```python
ds.sweep_names()
```

```python
sweep = ds.sweep('')
```

```python
f = sweep.variables['frequency'][:]
```

```python
s21 = sweep.variables['s21'][:].view(np.complex)
```

```python
fig, ax = plt.subplots()
ax.plot(f, 20*np.log10(np.abs(s21)))
ax.set_ylim(-90, -50)
```

```python
coarse = dataset.CoarseSweep(f, s21, coarse_resolution=0.125, n_subsweeps=16)
```

```python
n = 8
fig, ax = plt.subplots()
ax.plot(coarse.f[coarse.nth_subsweep(n)], abs(coarse.s21[coarse.nth_subsweep(n)]))
ax.set_ylim(0, 0.002)
```

```python
good = np.arange(9)
```

```python
fig, ax = plt.subplots()
for n in good:
    ax.plot(coarse.f[coarse.nth_subsweep(n)], 20*np.log10(abs(coarse.s21[coarse.nth_subsweep(n)])), '.k', markersize=0.2)
ax.set_ylim(-72, -58)
ax.set_xlim(10, 220)
#ax.set_ylim(-65, -58)
#ax.set_xlim(110, 220)
```

```python
fig, ax = plt.subplots()
for n in good:
    ax.plot(coarse.f[coarse.nth_subsweep(n)], 20*np.angle(coarse.s21[coarse.nth_subsweep(n)]), '.k', markersize=0.2)
#ax.set_ylim(-72, -58)
ax.set_xlim(10, 220)
```
