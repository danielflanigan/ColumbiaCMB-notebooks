```python
from scipy.signal import decimate, welch
from scipy.fftpack import fftshift, fftfreq
from kid_readout.utils import data_block, roach_interface, sweeps, data_file, readoutnc
from kid_readout.analysis import resonator, plot_resonator
```

```python
f0s = np.load('/home/data2/resonances/2014-11-17_0813f8_140825-3_fit_16.npy')
```

```python
nsamp = 2**20
step = 10
nstep = 1
f0binned = np.round(f0s*nsamp/512.0)*512.0/nsamp
offset_bins = np.arange(-(nstep+1),(nstep+1))*step
offsets = offset_bins*512.0/nsamp
```

```python
ri = roach_interface.RoachBaseband()
ri.set_dac_atten(39)
```

```python
measured_freqs = sweeps.prepare_sweep(ri, f0binned, offsets, nsamp=nsamp)
sweep_data = sweeps.do_prepared_sweep(ri, nchan_per_step=16, reads_per_step=1)
```

```python
df = data_file.DataFile(suffix='normalization_test')
df.log_hw_state(ri) # This is necessary for ReadoutNetCDF to work
df.add_sweep(sweep_data)
df.sync()
df.close()
```

```python
nc = readoutnc.ReadoutNetCDF(df.filename)
```

```python
sweep_data_freqs = sweep_data.freqs
sweep_data_means = sweep_data.data
sweep_data_errors = sweep_data.errors
nc_freqs = nc.sweeps[0].frequency
nc_means = nc.sweeps[0].s21
nc_errors = nc.sweeps[0].errors
```

```python
plt.plot(nc_freqs, abs(nc_means), '.')
plt.plot(sweep_data_freqs, abs(sweep_data_means), '.')
```

```python
plt.errorbar(nc_means.real, nc_means.imag, xerr=1e2*nc_errors.real, yerr=1e2*nc_errors.imag, fmt='.')
plt.errorbar(sweep_data_means.real, sweep_data_means.imag, xerr=1e2*sweep_data_errors.real, yerr=1e2*sweep_data_errors.imag, fmt='.')
```

```python
plt.plot(nc_freqs, nc_errors.real, '.')
plt.plot(sweep_data_freqs, sweep_data_errors.real, '.')
```

```python
plt.plot(nc_freqs, nc_errors.imag, '.')
plt.plot(sweep_data_freqs, sweep_data_errors.imag, '.')
```

```python
plt.plot(nc_freqs, abs(nc_errors), '.')
plt.plot(sweep_data_freqs, np.sqrt(2) * abs(sweep_data_errors), '.')
```
