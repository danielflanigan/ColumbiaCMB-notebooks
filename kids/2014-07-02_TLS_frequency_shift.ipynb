```python
from scipy.special import psi
from scipy.constants import c, h, k
```

The lowest value of $\frac{h f}{k T}$ achieved in our measurements was around $f = 77 \, \mathrm{MHz}$ and $T = 0.4 \, \mathrm{K}$:

```python
h * 77e6 / (k * 0.4)
```

The highest value achieved in Peter's measurements was at $f = 170\, \mathrm{MHz}$ and $T = 0.02 \, \mathrm{K}$:

```python
h * 170e6 / (k * 0.02)
```

Note that this is still far from saturation.

```python
xx = np.linspace(0.0001, 4, 10000)
digamma = np.real(psi(0.5 + xx / (2j * np.pi)))
log = -np.log(xx)
```

```python
fig, ax = plt.subplots()
ax.plot(xx,  digamma, label=r'$\operatorname{Re} \Psi \left(\frac{1}{2} + \frac{1}{2 \pi j} \frac{h f}{k T} \right)$')
ax.plot(xx, log, label=r'$-log \left( \frac{h f}{k T} \right)$')
ax.plot(xx, digamma + log, label=r'sum: $\frac{\pi}{F_{\mathrm{TLS}} \delta_0} x_{\mathrm{TLS}}$')
ax.set_xlabel(r'$\frac{h f}{k T}$')
ax.set_ylim(-3, 7)
ax.legend();
```
