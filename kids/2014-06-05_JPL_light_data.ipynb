All of the data shown here is extracted from least-square fits to the model
\begin{equation}
S_{21}(f) = A e^{-2 \pi j D f} \left( 1 - \frac{Q}{Q_e} \frac{1}{1 + 2 j Q (f - f_0) / f_0} \right),
\end{equation}
where $A$ is a complex scale factor, $D$ is the cable delay, $f$ is the probe tone frequency, $f_0$ is the resonant frequency, $Q$ is the loaded resonator quality factor, and $Q_e$ is the complex coupling quality factor as described in Khalil 2012. The loaded quality factor $Q$ is related to the internal quality factor $Q_i$ and the coupling quality factor $Q_e$ by
\begin{equation}
Q^{-1} = Q_i^{-1} + \operatorname{Re} \left( Q_e^{-1} \right).
\end{equation}

# Black body temperature sweeps

Plot black body temperature in kelvin versus fractional frequency shift $\delta x$ in parts per million for all 14 working resonators.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [1e6 * (r.f_0 / bb[n][0].f_0 - 1) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel('$10^6 \delta x$')
ax.set_ylim(-350, 50);
```

Plot black body temperature in kelvin versus internal quality factor $Q_i$ for all 14 working resonators. Note that the devices with higher $Q_i$ at the lowest optical load tend to have lower $Q_i$ at the highest optical load. Do we expect these curves to cross like this?

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [r.Q_i for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$Q_i$');
```

Plot black body temperature in kelvin versus shift in inverse internal quality factor times $10^6$ for all 14 working resonators.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [1e6 * (r.Q_i**-1 - bb[n][0].Q_i**-1) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylim(-1, 14);
```

Plot black body temperature in kelvin versus the absolute value of the complex coupling quality factor for all 14 working resonators. This quantity is not expected to vary significantly with changes in either black body temperature or bath temperature.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [abs(r.Q_e) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$|Q_e|$');
```

Plot the real part versus the imaginary part of the complex coupling quality factor.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.Q_e_real for r in bb[n]], [r.Q_e_imag for r in bb[n]], '.')
ax.set_xlabel(r'$\operatorname{Re} \, Q_e$')
ax.set_ylabel(r'$\operatorname{Im} \, Q_e$')
ax.set_xticks(np.linspace(0, 2e5, 5));
```

Plot black body temperature in kelvin versus the dimensionless coupling constant
$$\chi_c = \frac{4 Q_i Q_c}{(Q_i + Q_c)^2},$$
where $Q_c = \operatorname{Re} Q_e$ refers to the real part of the Khalil complex coupling quality factor.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [r.chi_c_real for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$\chi_c$')
ax.set_ylim(0.75, 1.02);
```

# Comparison of optical and thermal responses

For both the black body and bath temperature sweeps, plot fractional frequency shift versus inverse internal quality factor shift for resonator 0. This device has the lowest resonant frequency. The device temperature sweep data is fairly linear except at the lowest temperatures and the optical sweep data is very linear; minus two times the slope of this line equals $\beta$, the ratio of frequency response to disspation response.

The reference point for the shifts is at $T_{\mathrm{bath}} = 0.2 \, \mathrm{K}$ and $T_{\mathrm{bb}} = 4.2 \, \mathrm{K}$, since this point is common to both data sets.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot([1e6 * (r.Q_i**-1 - bath[0][0].Q_i**-1) for r in bath[0]],
        [1e6 * (r.f_0 / bath[0][0].f_0 - 1) for r in bath[0]],
        '.g', label='bath')
ax.plot([1e6 * (r.Q_i**-1 - bb[0][0].Q_i**-1) for r in bb[0]],
        [1e6 * (r.f_0 / bb[0][0].f_0 - 1) for r in bb[0]],
        '.b', label='black body')
ax.set_xlabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylabel('$10^6 \delta x$')
ax.set_xlim(-2, 22)
ax.legend();
```

Plot the same as above for resonator 13, which has the highest resonant frequency.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot([1e6 * (r.Q_i**-1 - bath[13][0].Q_i**-1) for r in bath[13]],
        [1e6 * (r.f_0 / bath[13][0].f_0 - 1) for r in bath[13]],
        '.g', label='bath')
ax.plot([1e6 * (r.Q_i**-1 - bb[13][0].Q_i**-1) for r in bb[13]],
        [1e6 * (r.f_0 / bb[13][0].f_0 - 1) for r in bb[13]],
        '.b', label='black body')
ax.set_xlabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylabel('$10^6 \delta x$')
ax.legend();
```

# Comparison to theory

Compare the expected value of $\beta$ from BCS theory to the value obtained from linear fits to the optical sweep data. Assume that the gap follows the BCS relation $2 \, \Delta_0 = 3.52 \, k_B T_c$. The solid curve below is the Mattis-Bardeen prediction and is not a fit to the data. 

```python
from physics.KIDs import noroozian
```

```python
T_bath = 0.2
T_c = 1.5
Delta = noroozian.BCS_Delta_0(T_c)
ff = np.linspace(bb[0][0].f_0, bb[13][0].f_0, 1e2)
# Calculate using frequency in Hz
theory_beta = noroozian.beta(1e6 * ff, T_bath, Delta)
f0 = [bb[n][-1].f_0 for n in range(n_kids)]
measured_beta = []
for n in range(n_kids):
    slope, intercept = np.polyfit([r.Q_i**-1 - bb[n][0].Q_i**-1 for r in bb[n]],
                                  [r.f_0 / bb[n][0].f_0 - 1 for r in bb[n]],
                                  1)
    measured_beta.append(-slope)
```

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot(ff, theory_beta, '-g', label='expected')
ax.plot(f0, measured_beta, '.b', label='measured')
ax.set_xlabel('frequency / [MHz]')
ax.set_ylabel(r'$\beta(f, T = 0.2 \, \mathrm{K}, T_c = 1.5 \, \mathrm{K})$')
ax.legend();
```
