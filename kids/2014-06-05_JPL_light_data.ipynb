```python
bb = data.JPL_2014_May_light_blackbody_temperature()
bath = data.JPL_2014_May_light_bath_temperature()
```

```python
n_kids = 14
```

# Black body temperature sweeps

Plot black body temperature in kelvin versus fractional frequency shift $\delta x$ in parts per million for all 14 working resonators.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [1e6 * (r.f_0 / bb[n][0].f_0 - 1) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel('$10^6 \delta x$')
ax.set_ylim(-350, 50);
```

Plot black body temperature in kelvin versus shift in inverse internal quality factor times $10^6$ for all 14 working resonators.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [1e6 * (r.Q_i**-1 - bb[n][0].Q_i**-1) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylim(-1, 14);
```

Plot black body temperature in kelvin versus the absolute value of the complex coupling quality factor for all 14 working resonators. This quantity is not expected to vary significantly with changes in either black body temperature or bath temperature.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [abs(r.Q_e) for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$|Q_c|$')
```

Plot the real part versus the imaginary part of the complex coupling quality factor.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.Q_e_real for r in bb[n]], [r.Q_e_imag for r in bb[n]], '.')
ax.set_xlabel('real part')
ax.set_ylabel('imaginary part')
ax.set_xticks(np.linspace(0, 2e5, 5));
```

Plot black body temperature in kelvin versus the dimensionless coupling constant
$$\chi_c = \frac{4 Q_i Q_c}{(Q_i + Q_c)^2},$$
where $Q_c$ refers to the real part of the Khalil complex coupling quality factor.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
for n in range(n_kids):
    ax.plot([r.T_bb for r in bb[n]], [r.chi_c_real for r in bb[n]], '-')
ax.set_xlabel('black body temperature / [K]')
ax.set_ylabel(r'$\chi_c$')
ax.set_ylim(0.75, 1.02);
```

# Comparison of optical and thermal responses

For both the black body and bath temperature sweeps, plot fractional frequency shift versus inverse internal quality factor shift for resonator 0. This device has the lowest resonant frequency. The device temperature sweep data is fairly linear except at the lowest temperatures and the optical sweep data is very linear; minus one times the slope of this line equals $\beta$, the ratio of frequency response to disspation response.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot([1e6 * (r.Q_i**-1 - bb[0][0].Q_i**-1) for r in bb[0]],
        [1e6 * (r.f_0 / bb[0][0].f_0 - 1) for r in bb[0]],
        '.b', label='black body')
ax.plot([1e6 * (r.Q_i**-1 - bath[0][0].Q_i**-1) for r in bath[0]],
        [1e6 * (r.f_0 / bath[0][0].f_0 - 1) for r in bath[0]],
        '.g', label='bath')
ax.set_xlabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylabel('$10^6 \delta x$')
ax.set_xlim(-2, 22)
ax.legend();
```

Plot the same as above for resonator 13, which has the highest resonant frequency.

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot([1e6 * (r.Q_i**-1 - bb[13][0].Q_i**-1) for r in bb[13]],
        [1e6 * (r.f_0 / bb[13][0].f_0 - 1) for r in bb[13]],
        '.b', label='black body')
ax.plot([1e6 * (r.Q_i**-1 - bath[13][0].Q_i**-1) for r in bath[13]],
        [1e6 * (r.f_0 / bath[13][0].f_0 - 1) for r in bath[13]],
        '.g', label='bath')
ax.set_xlabel(r'$10^6 \delta \left( Q_i^{-1} \right)$')
ax.set_ylabel('$10^6 \delta x$')
ax.legend();
```

# Comparison to theory

Compare the expected value of $\beta$ from BCS theory to the value obtained from linear fits to the optical sweep data. Assume that the gap follows the BCS relation $2 \, \Delta_0 = 3.52 \, k_B T_c$.

```python
from physics.KIDs import noroozian
```

```python
T_bath = 0.2
T_c = 1.5
Delta = noroozian.BCS_Delta_0(T_c)
ff = np.linspace(bb[0][0].f_0, bb[13][0].f_0, 1e2)
# Calculate using frequency in Hz
theory_beta = noroozian.beta(1e6 * ff, T_bath, Delta)
f0 = [bb[n][-1].f_0 for n in range(n_kids)]
measured_beta = []
for n in range(n_kids):
    p = np.polyfit([r.Q_i**-1 - bb[n][0].Q_i**-1 for r in bb[n]],
                    [r.f_0 / bb[n][0].f_0 - 1 for r in bb[n]],
                    1)
    measured_beta.append(-p[0])
```

```python
fig, ax = plt.subplots()
ax.set_frame_on(False)
ax.plot(ff, theory_beta, '-g', label='expected')
ax.plot(f0, measured_beta, '.b', label='measured')
ax.set_xlabel('frequency / [MHz]')
ax.set_ylabel(r'$\beta(f, T = 0.2 \, \mathrm{K}, T_c = 1.5 \, \mathrm{K})$')
ax.legend();
```
