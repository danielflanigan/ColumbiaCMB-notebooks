```python
Fs = 7 # in Hz; this is the mlab.psd default
N = 2**15
T = N / Fs
dt = T / N
df = Fs / N
```

```python
noise = np.random.randn(N)
```

```python
plt.plot(noise)
plt.xlabel('sample')
plt.ylabel('amplitude')
```

```python
Pr, f = mlab.psd(noise, Fs=Fs, window=mlab.window_none)
```

```python
plt.plot(f, Pr)
```

```python
np.var(noise)
```

```python
np.sum(Pr)
```

```python
np.sum(Pr) * df
```

```python
np.sqrt(N)
```

```python
np.sqrt(N/2) * np.sum(Pr) * df
```

```python
np.trapz(Pr, dx=df)
```

```python
np.sqrt(N/2) * np.trapz(Pr, dx=df)
```

```python
PN, fN = mlab.psd(noise, NFFT=N, Fs=Fs)
```

```python
plt.plot(fN, PN, f, Pr)
```
