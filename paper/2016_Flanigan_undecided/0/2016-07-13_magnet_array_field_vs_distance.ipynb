```python
# Raw data: jack height [mm], B_relative [uT].
points = [(60.0, 38.3),
          (81.0, 50.3),
          (89.5, 56.2),
          (101.0, 66.5),
          (110.5, 76.7),
          (120.0, 89.3),
          (129.0, 103.7),
          (139.5, 124.3),
          (150.0, 149.1),
          (160.0, 184.1),
          (170.0, 230.2),
          (180.5, 288.6)
        ]
jack_height_mm = np.array([p[0] for p in points])
B_relative_uT = np.array([p[1] for p in points])

probe_height_mm = 325 + 0.5  # The 0.5 is the distance from the probe end to the actual Hall sensor
block_thickness_mm = 25.5
magnet_thickness_mm = 3
z_mm = probe_height_mm - (jack_height_mm + block_thickness_mm + magnet_thickness_mm)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('magnet distance $z$ [mm]')
ax.set_ylabel('$B \; [\mu \mathrm{T}]$')
ax.plot(z_mm, B_relative_uT, '.')
```

```python
m, offset = np.polyfit(z_mm**-3, B_relative_uT, 1)
```
