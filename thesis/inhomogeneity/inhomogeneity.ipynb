```python
%matplotlib inline
import thesis
chapter = 'inhomogeneity'
save = True
```

```python
from scipy.constants import h
from thesis.constants import Delta, f_mc, f_1p, aluminum
v_F = aluminum['v_Fermi']
```

```python
E = Delta * (1 + np.linspace(1e-6, 0.2, 1000))
v_g = v_F * (1 - (Delta / E)**2)**(1/2)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('$E \; / \; \Delta$')
ax.set_ylabel('$v_\mathrm{g}(E) \; / \; v_\mathrm{F}$', rotation='horizontal', labelpad=25)
ax.axvline(1 + h * f_1p / Delta, color='gray', label='$f_\mathrm{1p}$')
ax.axvline(1 + h * f_mc / Delta, color='brown', label='$f_\mathrm{mc}$')
ax.plot(E / Delta, v_g / v_F, )
ax.set_xlim(0.98, (E / Delta).max())
ax.set_ylim(0, (v_g / v_F).max())
ax.legend()

if save:
    fig.savefig(thesis.full_filename(chapter, 'quasiparticle_group_velocity.pdf'))
```
