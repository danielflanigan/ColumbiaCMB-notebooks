```python
def dos_energy(E, Delta, Gamma):
    return np.real((E - 1j * Gamma) /
                   np.sqrt((E - 1j * Gamma)**2 - Delta**2))

def dos_gap(E, Delta0, Delta2):
    return np.real(E /
                   np.sqrt(E**2 - (Delta0 - 1j * Delta2)**2)) 
```

```python
fig, ax = plt.subplots()
#ax.set_xscale('log')
ax.set_yscale('log')
for broadening in (0.001, 0.002):
    Delta = 1
    EE = np.linspace(0, 2 * Delta, 10000)
    ax.plot(EE / Delta, dos_energy(E=EE, Delta=Delta, Gamma=Delta * broadening),
            label='$\Gamma/\Delta = {:.4f}$'.format(broadening))
    ax.plot(EE / Delta, dos_gap(E=EE, Delta0=Delta, Delta2=Delta * broadening),
            label='$\Delta_2 / \Delta_0 = {:.4f}$'.format(broadening))
ax.legend(fontsize='xx-small')    
#ax.set_xlim(1, 1.001)
```
