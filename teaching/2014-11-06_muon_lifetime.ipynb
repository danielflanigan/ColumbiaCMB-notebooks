Calculate the background rate:

```python
singles = 8.6 # in Hz
window = 1e-6 * (10 - 0.2) # in s
false_doubles = window * singles**2 # in s
elapsed_time = 6 * 24 * 3600 # about six days, in s
total_background_counts = elapsed_time * false_doubles
total_background_counts
```

```python
good = df[~df.too_fast & ~df.too_noisy]
good_counts, good_bins = np.histogram(good.dt, np.linspace(0.2, 10, 25))
corrected_counts = good_counts - total_background_counts / (good_bins.size - 1)
corrected_fracs = corrected_counts / corrected_counts.sum()
result = ml.fit(good_bins, corrected_fracs, ml.true_value, weights=corrected_fracs)
```

```python
fig = ml.plot_fit(good_bins, corrected_fracs, ml.model, result.params, good_counts.sum())
```

Calculate the expected measured lifetime; all times are in microseconds:

```python
R = 1.2766 # from the CMS measurement paper
f_p = R / (1 + R)
f_n = 1 / (1 + R)
tau_p = ml.tau
tau_n = 2.03 # from the Suzuki paper
tau_expected = f_p * tau_p + f_n * tau_n
tau_expected
```
